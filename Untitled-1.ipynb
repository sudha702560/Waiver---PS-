```python
"""
=============================================================
WAIVER OPTIMIZER — MULTI-START SLSQP
=============================================================

"""

import pandas as pd
import numpy as np
from scipy.optimize import minimize
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from openpyxl import Workbook
from openpyxl.styles import Font, PatternFill, Alignment, Border, Side
from openpyxl.utils import get_column_letter
import time
import warnings
warnings.filterwarnings('ignore')

# ─────────────────────────────────────────────────────────
# PATHS  ← update these for your machine
# ─────────────────────────────────────────────────────────
INPUT_CSV = r"C:\Users\intern-Sudhashri\Desktop\WAVIER\waiver_enriched_20000.csv"
OUT_CSV   = r"C:\Users\intern-Sudhashri\Desktop\WAVIER\waiver_output_MultiStart_W005.csv"
OUT_EXCEL = r"C:\Users\intern-Sudhashri\Desktop\WAVIER\waiver_results_MultiStart_W005.xlsx"
# ─────────────────────────────────────────────────────────

# ══════════════════════════════════════════════════════════
# KEY PARAMETER — minimum weight value
# All 15 weights must be >= this value
# ══════════════════════════════════════════════════════════
W_MIN = 0.05   # weights > 0.05

t0 = time.time()

# =============================================================
# 1. LOAD DATA
# =============================================================
df = pd.read_csv(INPUT_CSV).copy()
n  = len(df)
print(f"[1] Loaded {n} rows x {df.shape[1]} cols")
np.random.seed(42)

# =============================================================
# 2. ENRICH DATA
# =============================================================
dlr_mask = df["Lead Source"].astype(str).str.strip().str.lower() == "dealership"
dlr_idx  = np.where(dlr_mask)[0]
n_high   = int(len(dlr_idx) * 0.60)
n_low    = len(dlr_idx) - n_high
hi_idx   = np.random.choice(dlr_idx, n_high, replace=False)
lo_idx   = np.setdiff1d(dlr_idx, hi_idx)

dq_new = df["Dealership Quality (DQ)"].values.astype(float).copy()
dq_new[hi_idx] = np.random.uniform(80, 100, n_high)
dq_new[lo_idx] = np.random.uniform(0,  79,  n_low)
df["Dealership Quality (DQ)"] = dq_new

lf_new = df["Lead Flow (LF%)"].values.astype(float).copy()
dp_new = df["Dealer Payout% (DP%)"].values.astype(float).copy()
lf_new[hi_idx] = np.random.uniform(0, 100, n_high)
dp_new[hi_idx] = np.random.uniform(0,   3, n_high)
df["Lead Flow (LF%)"]      = lf_new
df["Dealer Payout% (DP%)"] = dp_new

_rel_v  = df["Relationship with Company"].astype(str).values
_dpd_v  = df["Average Days Past Due (DPD) - Past 6 Months"].astype(float).values
_wvr_v  = df["Average Waiver Received Previously - Same Product"].astype(float).values
_prod_v = df["Previous Loan Product"].astype(str).str.strip().str.lower().values

prior_idx = np.where((_rel_v=="Existing Customer")&(_dpd_v<=5)&(_wvr_v>0))[0]
wvr_new   = _wvr_v.copy()
p4 = len(prior_idx)//4
wvr_new[prior_idx[:p4]]       = np.random.uniform(0.01,0.49,p4)
wvr_new[prior_idx[p4:2*p4]]   = np.random.uniform(0.50,0.99,p4)
wvr_new[prior_idx[2*p4:3*p4]] = np.random.uniform(1.00,1.49,p4)
wvr_new[prior_idx[3*p4:]]     = np.random.uniform(1.50,2.00,len(prior_idx)-3*p4)
df["Average Waiver Received Previously - Same Product"] = wvr_new

DOH = "Difference between Offered & Historical Interest Rate (Δoh)"
nps_idx = np.where(
    (_rel_v=="Existing Customer")&(_dpd_v<=5)&
    (wvr_new==0)&(_prod_v=="same product")
)[0]
doh_new = df[DOH].astype(float).values.copy()
p5 = len(nps_idx)//5
doh_new[nps_idx[:p5]]       = np.random.uniform(-2.0, 0.00,p5)
doh_new[nps_idx[p5:2*p5]]   = np.random.uniform( 0.01,0.49,p5)
doh_new[nps_idx[2*p5:3*p5]] = np.random.uniform( 0.50,0.99,p5)
doh_new[nps_idx[3*p5:4*p5]] = np.random.uniform( 1.00,1.49,p5)
doh_new[nps_idx[4*p5:]]     = np.random.uniform( 1.50,3.00,len(nps_idx)-4*p5)
df[DOH] = doh_new

print(f"[2] Data enriched in {round(time.time()-t0,2)}s")

# =============================================================
# 3. TARGETS & RATES
# =============================================================
y1  = df["Actual Waiver Granted (Interest Rate Waiver)"].fillna(0).values.astype(np.float32)
y2  = df["Actual Waiver Granted (Admin & Processing Fee Waiver)"].fillna(0).values.astype(np.float32)
y3  = df["Actual Waiver Granted (Sourcing Fee Waiver)"].fillna(0).values.astype(np.float32)
Ro  = df["Interest Rate Offered (Ro)"].fillna(0).values.astype(np.float32)
Rpo = df["Admin & Processing Fee Offered (Rpo)"].fillna(0).values.astype(np.float32)
Rso = df["Sourcing Fee Offered (Rso)"].fillna(0).values.astype(np.float32)

idx_all   = np.arange(n)
train_idx, test_idx = train_test_split(idx_all, test_size=0.20, random_state=42)
print(f"[3] Train: {len(train_idx)}, Test: {len(test_idx)}")

# =============================================================
# 4. CONDITIONS
# =============================================================
A_names      = ["A1","A2"]
A_conditions = ["Usage Type = Personal","Usage Type = Commercial"]

B_names      = [f"B{i}" for i in range(1,13)]
B_conditions = [
    "New Customer",
    "Existing Customer, Bad Repayment Record (avg DPD last 6M > 5)",
    "Existing Customer, Good Repayment, Received Prior Waiver, Bn ∈ (0, 0.5)",
    "Existing Customer, Good Repayment, Received Prior Waiver, Bn ∈ [0.5, 1.0)",
    "Existing Customer, Good Repayment, Received Prior Waiver, Bn ∈ [1.0, 1.5)",
    "Existing Customer, Good Repayment, Received Prior Waiver, Bn ∈ [1.5, ∞)",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product | Δoh ≤ 0",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product | Δoh ∈ (0, 0.5)",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product | Δoh ∈ [0.5, 1.0)",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product | Δoh ∈ [1.0, 1.5)",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product | Δoh ≥ 1.5",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Financing Different Product",
]

C_names      = [f"C{i}" for i in range(1,39)]
C_conditions = ["Source ≠ Dealership","Source = Dealership, DQ < 80%"]
for lf_lbl in ["LF < 10%","10% ≤ LF < 20%","20% ≤ LF < 30%",
                "30% ≤ LF < 40%","40% ≤ LF < 50%","LF ≥ 50%"]:
    for dp_lbl in ["DP < 0.3%","0.3% ≤ DP < 0.6%","0.6% ≤ DP < 0.9%",
                   "0.9% ≤ DP < 1.2%","1.2% ≤ DP < 1.5%","DP ≥ 1.5%"]:
        C_conditions.append(f"Source = Dealership, DQ ≥ 80%, {lf_lbl}, {dp_lbl}")

D_names      = ["D1","D2","D3","D4"]
D_conditions = [
    "Branch Ahead of Profitability Target & Behind Disbursal Target",
    "Branch Ahead of Profitability Target & Ahead of Disbursal Target",
    "Branch Behind Profitability Target & Behind Disbursal Target",
    "Branch Behind Profitability Target & Ahead of Disbursal Target",
]

E_names      = [f"E{i}" for i in range(1,6)]
E_conditions = [
    "Category 1 – Tier 1 Urban","Category 2 – Tier 1 Rural",
    "Category 3 – Tier 2 Urban","Category 4 – Tier 2 Rural",
    "Category 5 – Tier 3 Rural & Others",
]

# =============================================================
# 5. BUCKET ASSIGNMENT (vectorized)
# =============================================================
A_idx = np.where(df["Usage Type"].eq("Personal"),0,1).astype(np.int32)

_rel  = df["Relationship with Company"].astype(str).values
_dpd  = df["Average Days Past Due (DPD) - Past 6 Months"].astype(float).values
_wvr  = df["Average Waiver Received Previously - Same Product"].astype(float).values
_doh  = df[DOH].astype(float).values
_prod = df["Previous Loan Product"].astype(str).str.strip().str.lower().values

is_new  = (_rel=="New Customer")
is_bad  = (~is_new)&(_dpd>5)
is_good = (~is_new)&(_dpd<=5)
prior   = is_good&(_wvr>0)
nps     = is_good&(_wvr==0)&(_prod=="same product")
diffp   = is_good&(_wvr==0)&(_prod=="different")

B_idx = np.zeros(n,dtype=np.int32)
B_idx[is_new]=0;  B_idx[is_bad]=1
B_idx[prior&(_wvr<0.5)]=2;  B_idx[prior&(_wvr>=0.5)&(_wvr<1.0)]=3
B_idx[prior&(_wvr>=1.0)&(_wvr<1.5)]=4;  B_idx[prior&(_wvr>=1.5)]=5
B_idx[nps&(_doh<=0)]=6;     B_idx[nps&(_doh>0)&(_doh<0.5)]=7
B_idx[nps&(_doh>=0.5)&(_doh<1.0)]=8
B_idx[nps&(_doh>=1.0)&(_doh<1.5)]=9
B_idx[nps&(_doh>=1.5)]=10;  B_idx[diffp]=11

_src   = df["Lead Source"].astype(str).str.strip().str.lower().values
_dq    = df["Dealership Quality (DQ)"].astype(float).values
_lf    = df["Lead Flow (LF%)"].astype(float).values
_dp    = df["Dealer Payout% (DP%)"].astype(float).values
is_dlr = (_src=="dealership")
dlr_ok = is_dlr&(_dq>=80)
lf_grp = np.where(_lf<10,0,np.where(_lf<20,1,np.where(_lf<30,2,
         np.where(_lf<40,3,np.where(_lf<50,4,5))))).astype(np.int32)
dp_grp = np.where(_dp<0.3,0,np.where(_dp<0.6,1,np.where(_dp<0.9,2,
         np.where(_dp<1.2,3,np.where(_dp<1.5,4,5))))).astype(np.int32)
C_idx = np.zeros(n,dtype=np.int32)
C_idx[~is_dlr]=0;  C_idx[is_dlr&(_dq<80)]=1
C_idx[dlr_ok] = 2+lf_grp[dlr_ok]*6+dp_grp[dlr_ok]

_profit = df["Tracking towards profitability target"].astype(str).values
_disb   = df["Tracking towards disbursal target"].astype(str).values
D_idx   = np.full(n,2,dtype=np.int32)
D_idx[(_profit=="Ahead")&(_disb=="Behind")]=0
D_idx[(_profit=="Ahead")&(_disb=="Ahead")] =1
D_idx[(_profit=="Behind")&(_disb=="Ahead")]=3

_ecat = df["Branch Competitive Category"].astype(str).str.strip().values
E_idx = np.zeros(n,dtype=np.int32)
for digit,ei in [("1",0),("2",1),("3",2),("4",3),("5",4)]:
    E_idx[pd.Series(_ecat).str.contains(digit,regex=False).values]=ei

print("[5] Bucket assignment done")

# =============================================================
# 6. ONE-HOT MATRICES
# =============================================================
def make_oh(ia,nb,rows):
    M=np.zeros((len(rows),nb),dtype=np.float32)
    M[np.arange(len(rows)),ia[rows]]=1.0
    return M

tr = train_idx
OA=make_oh(A_idx,2,tr);   OA_all=make_oh(A_idx,2,idx_all)
OB=make_oh(B_idx,12,tr);  OB_all=make_oh(B_idx,12,idx_all)
OC=make_oh(C_idx,38,tr);  OC_all=make_oh(C_idx,38,idx_all)
OD=make_oh(D_idx,4,tr);   OD_all=make_oh(D_idx,4,idx_all)
OE=make_oh(E_idx,5,tr);   OE_all=make_oh(E_idx,5,idx_all)

y1_tr=y1[tr]; y2_tr=y2[tr]; y3_tr=y3[tr]
s1=0.1*Ro[tr]; s2=0.1*Rpo[tr]; s3=0.1*Rso[tr]
print(f"[6] One-hot matrices built in {round(time.time()-t0,2)}s")

# =============================================================
# 7. OBJECTIVE FUNCTION
# =============================================================
def objective(p):
    A_=p[0:2].astype(np.float32);  B_=p[2:14].astype(np.float32)
    C_=p[14:52].astype(np.float32); D_=p[52:56].astype(np.float32)
    E_=p[56:61].astype(np.float32)
    w1_=p[61:66]; w2_=p[66:71]; w3_=p[71:76]
    Av=OA@A_; Bv=OB@B_; Cv=OC@C_; Dv=OD@D_; Ev=OE@E_
    sc1=w1_[0]*Av+w1_[1]*Bv+w1_[2]*Cv+w1_[3]*Dv+w1_[4]*Ev
    sc2=w2_[0]*Av+w2_[1]*Bv+w2_[2]*Cv+w2_[3]*Dv+w2_[4]*Ev
    sc3=w3_[0]*Av+w3_[1]*Bv+w3_[2]*Cv+w3_[3]*Dv+w3_[4]*Ev
    r1=y1_tr-sc1*s1; r2=y2_tr-sc2*s2; r3=y3_tr-sc3*s3
    return float(r1@r1+r2@r2+r3@r3+0.001*np.dot(p,p))

# =============================================================
# 8. CONSTRAINTS
# ── CHANGE: Added weight > W_MIN constraints for all 15 weights
# =============================================================
cons = [
    # Weights sum = 1
    {"type":"eq",   "fun": lambda p: np.sum(p[61:66])-1},
    {"type":"eq",   "fun": lambda p: np.sum(p[66:71])-1},
    {"type":"eq",   "fun": lambda p: np.sum(p[71:76])-1},

    # Business rules
    {"type":"ineq", "fun": lambda p: p[0]-p[1]-0.5},    # A1 > A2
    {"type":"eq",   "fun": lambda p: p[2]-5},            # B1 = 5
    {"type":"eq",   "fun": lambda p: p[3]},              # B2 = 0
    {"type":"ineq", "fun": lambda p: p[5]-p[4]-0.1},    # B4 > B3
    {"type":"ineq", "fun": lambda p: p[6]-p[5]-0.1},    # B5 > B4
    {"type":"ineq", "fun": lambda p: p[7]-p[6]-0.1},    # B6 > B5
    {"type":"eq",   "fun": lambda p: p[13]-5},           # B12 = 5
    {"type":"eq",   "fun": lambda p: p[14]-5},           # C1 = 5
    {"type":"eq",   "fun": lambda p: p[15]},             # C2 = 0
    {"type":"ineq", "fun": lambda p: p[52]-p[53]-0.1},  # D1 > D2
    {"type":"ineq", "fun": lambda p: p[53]-p[54]-0.1},  # D2 > D3
    {"type":"ineq", "fun": lambda p: p[54]-p[55]-0.1},  # D3 > D4
    {"type":"eq",   "fun": lambda p: p[55]},             # D4 = 0
    {"type":"ineq", "fun": lambda p: p[56]-p[57]-0.1},  # E1 > E2
    {"type":"ineq", "fun": lambda p: p[57]-p[58]-0.1},  # E2 > E3
    {"type":"ineq", "fun": lambda p: p[58]-p[59]-0.1},  # E3 > E4
    {"type":"ineq", "fun": lambda p: p[59]-p[60]-0.1},  # E4 > E5
    {"type":"eq",   "fun": lambda p: p[60]-1},           # E5 = 1

    # ── WEIGHTS > W_MIN (0.05) — all 15 weights ──────────────
    # W1 weights (p[61] to p[65])
    {"type":"ineq", "fun": lambda p: p[61]-W_MIN},  # w1A >= 0.05
    {"type":"ineq", "fun": lambda p: p[62]-W_MIN},  # w1B >= 0.05
    {"type":"ineq", "fun": lambda p: p[63]-W_MIN},  # w1C >= 0.05
    {"type":"ineq", "fun": lambda p: p[64]-W_MIN},  # w1D >= 0.05
    {"type":"ineq", "fun": lambda p: p[65]-W_MIN},  # w1E >= 0.05
    # W2 weights (p[66] to p[70])
    {"type":"ineq", "fun": lambda p: p[66]-W_MIN},  # w2A >= 0.05
    {"type":"ineq", "fun": lambda p: p[67]-W_MIN},  # w2B >= 0.05
    {"type":"ineq", "fun": lambda p: p[68]-W_MIN},  # w2C >= 0.05
    {"type":"ineq", "fun": lambda p: p[69]-W_MIN},  # w2D >= 0.05
    {"type":"ineq", "fun": lambda p: p[70]-W_MIN},  # w2E >= 0.05
    # W3 weights (p[71] to p[75])
    {"type":"ineq", "fun": lambda p: p[71]-W_MIN},  # w3A >= 0.05
    {"type":"ineq", "fun": lambda p: p[72]-W_MIN},  # w3B >= 0.05
    {"type":"ineq", "fun": lambda p: p[73]-W_MIN},  # w3C >= 0.05
    {"type":"ineq", "fun": lambda p: p[74]-W_MIN},  # w3D >= 0.05
    {"type":"ineq", "fun": lambda p: p[75]-W_MIN},  # w3E >= 0.05
]

# Bounds — bucket values 0-10, weights 0.05-1
bounds = [(0,10)]*61 + [(W_MIN,1)]*15

# =============================================================
# 9. MULTIPLE STARTING POINTS
# =============================================================
def make_start(a1, a2, seed):
    np.random.seed(seed)
    x = np.zeros(76)
    x[0]=a1;   x[1]=a2
    x[2]=5.0;  x[3]=0.0
    x[4:8]   = np.sort(np.random.uniform(0.5,4.0,4))
    x[8:13]  = np.sort(np.random.uniform(0.5,5.0,5))
    x[13]    = 5.0
    x[14]    = 5.0;  x[15]=0.0
    x[16:52] = np.random.uniform(0.5,4.5,36)
    x[52:55] = np.sort(np.random.uniform(0.5,4.0,3))[::-1]
    x[55]    = 0.0
    x[56:60] = np.sort(np.random.uniform(1.5,5.0,4))[::-1]
    x[60]    = 1.0
    # Weights start at 0.2 (= 1/5), all equal and above W_MIN
    x[61:76] = 0.2
    return x

starting_points = [
    make_start(7.0, 4.0, 42),
    make_start(5.0, 2.0, 10),
    make_start(9.0, 6.0, 20),
    make_start(3.0, 1.5, 30),
    make_start(6.0, 3.0, 40),
]

# =============================================================
# 10. MULTI-START OPTIMIZE
# =============================================================
print(f"\n[10] Multi-Start SLSQP — {len(starting_points)} runs | W_MIN={W_MIN}\n")

best_result = None
best_loss   = float('inf')
run_log     = []
t_opt = time.time()

for i, x0 in enumerate(starting_points):
    print(f"  Run {i+1}/{len(starting_points)} | A1={x0[0]:.1f}  A2={x0[1]:.1f}")
    res = minimize(
        objective, x0,
        method="SLSQP",
        bounds=bounds,
        constraints=cons,
        options={"maxiter":200,"ftol":1e-5}
    )
    print(f"  Loss={res.fun:.6f} | {res.message}")
    run_log.append({
        "run":    i+1,
        "A1":     round(float(res.x[0]),4),
        "A2":     round(float(res.x[1]),4),
        "loss":   round(res.fun,6),
        "status": res.message
    })
    if res.fun < best_loss:
        best_loss   = res.fun
        best_result = res
        print(f"  ✓ NEW BEST\n")
    else:
        print()

print(f"\n  Run summary:")
print(f"  {'Run':>4}  {'A1':>8}  {'A2':>8}  {'Loss':>12}  Best?")
print(f"  {'─'*50}")
for log in run_log:
    mark = " ← BEST" if log["loss"]==best_loss else ""
    print(f"  {log['run']:>4}  {log['A1']:>8}  {log['A2']:>8}  "
          f"{log['loss']:>12.6f}{mark}")
print(f"\n  Best loss = {best_loss:.6f}")
print(f"  Time      = {round(time.time()-t_opt,2)}s")

p  = best_result.x
A  = p[0:2];   B  = p[2:14];  C  = p[14:52]
D  = p[52:56]; E  = p[56:61]
w1 = p[61:66]; w2 = p[66:71]; w3 = p[71:76]

# Verify weights > W_MIN
print(f"\n  Weight verification (all must be >= {W_MIN}):")
print(f"  W1={np.round(w1,4)}  min={min(w1):.4f}  "
      f"{'✓' if min(w1)>=W_MIN-1e-9 else '✗'}")
print(f"  W2={np.round(w2,4)}  min={min(w2):.4f}  "
      f"{'✓' if min(w2)>=W_MIN-1e-9 else '✗'}")
print(f"  W3={np.round(w3,4)}  min={min(w3):.4f}  "
      f"{'✓' if min(w3)>=W_MIN-1e-9 else '✗'}")

# =============================================================
# 11. PREDICT
# =============================================================
Av=OA_all@A; Bv=OB_all@B; Cv=OC_all@C; Dv=OD_all@D; Ev=OE_all@E
sc1=w1[0]*Av+w1[1]*Bv+w1[2]*Cv+w1[3]*Dv+w1[4]*Ev
sc2=w2[0]*Av+w2[1]*Bv+w2[2]*Cv+w2[3]*Dv+w2[4]*Ev
sc3=w3[0]*Av+w3[1]*Bv+w3[2]*Cv+w3[3]*Dv+w3[4]*Ev

pred_w1=sc1*0.1*Ro; pred_w2=sc2*0.1*Rpo; pred_w3=sc3*0.1*Rso
actual_total=y1+y2+y3; pred_total=pred_w1+pred_w2+pred_w3

def metrics(a,q):
    mse=float(mean_squared_error(a,q))
    return round(mse,6),round(float(np.sqrt(mse)),6),round(float(np.mean(np.abs(a-q))),6)

tr_m=metrics(actual_total[train_idx],pred_total[train_idx])
te_m=metrics(actual_total[test_idx], pred_total[test_idx])
print(f"\n[11] Train → MSE={tr_m[0]}  RMSE={tr_m[1]}  MAE={tr_m[2]}")
print(f"     Test  → MSE={te_m[0]}  RMSE={te_m[1]}  MAE={te_m[2]}")

# =============================================================
# 12. BUILD CSV
# ── CHANGE: Residual W1/W2/W3/Total columns REMOVED
# =============================================================
split_col = np.where(np.isin(idx_all,train_idx),"Train","Test")

out = pd.DataFrame({
    "Customer":        df["Name of Customer"].values,
    "A Category":      [A_names[i] for i in A_idx],
    "A Condition":     [A_conditions[i] for i in A_idx],
    "A Row Value":     np.round(A[A_idx],4),
    "B Category":      [B_names[i] for i in B_idx],
    "B Condition":     [B_conditions[i] for i in B_idx],
    "B Row Value":     np.round(B[B_idx],4),
    "C Category":      [C_names[i] for i in C_idx],
    "C Condition":     [C_conditions[i] for i in C_idx],
    "C Row Value":     np.round(C[C_idx],4),
    "D Category":      [D_names[i] for i in D_idx],
    "D Condition":     [D_conditions[i] for i in D_idx],
    "D Row Value":     np.round(D[D_idx],4),
    "E Category":      [E_names[i] for i in E_idx],
    "E Condition":     [E_conditions[i] for i in E_idx],
    "E Row Value":     np.round(E[E_idx],4),
    "W1A":round(w1[0],4),"W1B":round(w1[1],4),"W1C":round(w1[2],4),
    "W1D":round(w1[3],4),"W1E":round(w1[4],4),
    "W2A":round(w2[0],4),"W2B":round(w2[1],4),"W2C":round(w2[2],4),
    "W2D":round(w2[3],4),"W2E":round(w2[4],4),
    "W3A":round(w3[0],4),"W3B":round(w3[1],4),"W3C":round(w3[2],4),
    "W3D":round(w3[3],4),"W3E":round(w3[4],4),
    "Score W1":     np.round(sc1,4),
    "Score W2":     np.round(sc2,4),
    "Score W3":     np.round(sc3,4),
    "Ro":           np.round(Ro,4),
    "Rpo":          np.round(Rpo,4),
    "Rso":          np.round(Rso,4),
    "Actual W1":    np.round(y1,4),
    "Actual W2":    np.round(y2,4),
    "Actual W3":    np.round(y3,4),
    "Actual Total": np.round(actual_total,4),
    "Predicted W1": np.round(pred_w1,4),
    "Predicted W2": np.round(pred_w2,4),
    "Predicted W3": np.round(pred_w3,4),
    "Predicted Total": np.round(pred_total,4),
    # ── Residual W1, W2, W3, Total REMOVED ──
    "Split":        split_col,
})

out.to_csv(OUT_CSV,index=False)
print(f"\n[12] CSV saved → {len(out)} rows x {len(out.columns)} cols")

# =============================================================
# 13. BUILD EXCEL — same 5-sheet format as original
# =============================================================
print("[13] Building Excel workbook...")

cnt_A=pd.Series([A_names[i] for i in A_idx]).value_counts()
cnt_B=pd.Series([B_names[i] for i in B_idx]).value_counts()
cnt_C=pd.Series([C_names[i] for i in C_idx]).value_counts()
cnt_D=pd.Series([D_names[i] for i in D_idx]).value_counts()
cnt_E=pd.Series([E_names[i] for i in E_idx]).value_counts()

all_groups=[("A",A_names,A_conditions,A,cnt_A),
            ("B",B_names,B_conditions,B,cnt_B),
            ("C",C_names,C_conditions,C,cnt_C),
            ("D",D_names,D_conditions,D,cnt_D),
            ("E",E_names,E_conditions,E,cnt_E)]

HDR=PatternFill("solid",start_color="1F4E79")
GF={"A":PatternFill("solid",start_color="4472C4"),
    "B":PatternFill("solid",start_color="375623"),
    "C":PatternFill("solid",start_color="843C0C"),
    "D":PatternFill("solid",start_color="7B3F00"),
    "E":PatternFill("solid",start_color="4B0082")}
AF={"A":PatternFill("solid",start_color="D6E4F7"),
    "B":PatternFill("solid",start_color="E2EFDA"),
    "C":PatternFill("solid",start_color="FCE4D6"),
    "D":PatternFill("solid",start_color="F4CCAC"),
    "E":PatternFill("solid",start_color="E8D5F5")}
WH =PatternFill("solid",start_color="FFFFFF")
GRN=PatternFill("solid",start_color="E2EFDA")
thin=Side(style="thin",color="BFBFBF")
BS  =Border(left=thin,right=thin,top=thin,bottom=thin)

def hc(ws,r,c,v,w=None):
    cl=ws.cell(r,c,v)
    cl.font=Font(bold=True,color="FFFFFF",size=11,name="Arial")
    cl.fill=HDR; cl.border=BS
    cl.alignment=Alignment(horizontal="center",vertical="center",wrap_text=True)
    if w: ws.column_dimensions[get_column_letter(c)].width=w

def dc(ws,r,c,v,fill=None,bold=False,align="left",fmt=None,fc="000000"):
    cl=ws.cell(r,c,v)
    cl.font=Font(bold=bold,size=10,name="Arial",color=fc)
    cl.fill=fill or WH; cl.border=BS
    cl.alignment=Alignment(horizontal=align,vertical="center",wrap_text=True)
    if fmt: cl.number_format=fmt

wb=Workbook()

# ── Sheet 1: Customer Output ──────────────────────────────
ws1=wb.active; ws1.title="Customer Output"
ws1.freeze_panes="B2"; ws1.row_dimensions[1].height=30
for ci,col in enumerate(out.columns,1):
    hc(ws1,1,ci,col,w=max(12,min(35,len(col)+2)))
for ri,row in enumerate(out.itertuples(index=False),2):
    fill=AF["A"] if ri%2==0 else WH
    for ci,val in enumerate(row,1):
        dc(ws1,ri,ci,val,fill=fill,align="center")
    # Color Split column green/red
    split_c=list(out.columns).index("Split")+1
    ws1.cell(ri,split_c).fill = (
        PatternFill("solid",start_color="E2EFDA")
        if row[split_c-1]=="Train"
        else PatternFill("solid",start_color="FFDADA")
    )
print("     Sheet 1 (Customer Output) done")

# ── Sheet 2: Condition Summary ────────────────────────────
ws2=wb.create_sheet("Condition Summary")
ws2.freeze_panes="A2"; ws2.row_dimensions[1].height=35
for ci,(h,w) in enumerate(zip(
    ["Group","State","Optimized Value","Condition",
     "No. of Customers","% of Total"],
    [8,8,16,90,20,14]),1):
    hc(ws2,1,ci,h,w)
rn=2
for grp,names,conds,vals,cnt_ser in all_groups:
    for k,(nm,cond,val) in enumerate(zip(names,conds,vals)):
        cnt=int(cnt_ser.get(nm,0)); f=AF[grp] if k%2==0 else WH
        dc(ws2,rn,1,grp,fill=GF[grp],bold=True,align="center",fc="FFFFFF")
        dc(ws2,rn,2,nm,fill=f,bold=True,align="center")
        dc(ws2,rn,3,round(float(val),4),fill=f,align="center",fmt="0.0000")
        dc(ws2,rn,4,cond,fill=f)
        dc(ws2,rn,5,cnt,fill=f,align="center",fmt="#,##0")
        dc(ws2,rn,6,round(cnt/n,5),fill=f,align="center",fmt="0.00%")
        if cnt==0:
            ws2.cell(rn,5).font=Font(bold=True,color="FF0000",size=10,name="Arial")
        rn+=1
for ci,val in enumerate(["","TOTAL","","",n,1.0],1):
    c=ws2.cell(rn,ci,val)
    c.font=Font(bold=True,color="FFFFFF",size=11,name="Arial")
    c.fill=HDR; c.border=BS
    c.alignment=Alignment(horizontal="center",vertical="center")
ws2.cell(rn,5).number_format="#,##0"
ws2.cell(rn,6).number_format="0.00%"
print("     Sheet 2 (Condition Summary) done")

# ── Sheet 3: Weights ──────────────────────────────────────
ws3=wb.create_sheet("Weights"); ws3.row_dimensions[1].height=30
for ci,(h,w) in enumerate(zip(
    ["Dimension","W1 (Interest Rate)","W2 (Admin Fee)","W3 (Sourcing Fee)"],
    [28,20,20,20]),1):
    hc(ws3,1,ci,h,w)

wt_rows=[
    ("wA – Usage Type",         w1[0],w2[0],w3[0]),
    ("wB – Customer / Repayment",w1[1],w2[1],w3[1]),
    ("wC – Dealership / Lead Source",w1[2],w2[2],w3[2]),
    ("wD – Branch Performance",  w1[3],w2[3],w3[3]),
    ("wE – Geography",           w1[4],w2[4],w3[4]),
    ("SUM",                      sum(w1),sum(w2),sum(w3)),
]

for ri,(lbl,v1,v2,v3) in enumerate(wt_rows,2):
    is_s=(lbl=="SUM")
    f=HDR if is_s else (AF["B"] if ri%2==0 else WH)
    fc="FFFFFF" if is_s else "000000"
    for ci,val in enumerate(
        [lbl,round(float(v1),4),round(float(v2),4),round(float(v3),4)],1):
        c=ws3.cell(ri,ci,val)
        c.font=Font(bold=is_s,color=fc,size=10,name="Arial")
        c.fill=f; c.border=BS
        c.alignment=Alignment(
            horizontal="center" if ci>1 else "left",
            vertical="center")
        if ci>1: c.number_format="0.0000"
print("     Sheet 3 (Weights) done")

# ── Sheet 4: Validation ───────────────────────────────────
ws4=wb.create_sheet("Validation"); ws4.row_dimensions[1].height=30
for ci,(h,w) in enumerate(zip(["Metric","Train","Test"],[15,18,18]),1):
    hc(ws4,1,ci,h,w)
for ri,(m,tv,te) in enumerate(zip(["MSE","RMSE","MAE"],tr_m,te_m),2):
    f=AF["A"] if ri%2==0 else WH
    dc(ws4,ri,1,m,fill=f,bold=True)
    dc(ws4,ri,2,tv,fill=f,align="center",fmt="0.000000")
    dc(ws4,ri,3,te,fill=f,align="center",fmt="0.000000")
print("     Sheet 4 (Validation) done")

# ── Sheet 5: Multi-Start Log ──────────────────────────────
ws5=wb.create_sheet("Multi-Start Log"); ws5.row_dimensions[1].height=30
for ci,(h,w) in enumerate(zip(
    ["Run","Starting A1","Starting A2","Final Loss","Status","Best?"],
    [8,16,16,18,40,12]),1):
    hc(ws5,1,ci,h,w)
for ri,log in enumerate(run_log,2):
    is_best=(log["loss"]==best_loss)
    f=GRN if is_best else WH
    dc(ws5,ri,1,log["run"],  fill=f,align="center",bold=is_best)
    dc(ws5,ri,2,log["A1"],   fill=f,align="center",fmt="0.0000")
    dc(ws5,ri,3,log["A2"],   fill=f,align="center",fmt="0.0000")
    dc(ws5,ri,4,log["loss"], fill=f,align="center",fmt="0.000000")
    dc(ws5,ri,5,log["status"],fill=f)
    dc(ws5,ri,6,"✓ BEST" if is_best else "",
       fill=f,align="center",bold=True,
       fc="085041" if is_best else "000000")
print("     Sheet 5 (Multi-Start Log) done")

wb.save(OUT_EXCEL)
print(f"\n[13] Excel saved → {OUT_EXCEL}")

# =============================================================
# 14. FINAL VERIFICATION
# =============================================================
print(f"\n{'='*65}")
print("FINAL VERIFICATION")
print(f"{'='*65}")
all_ok=True
for grp,names,_,_,cnt_ser in all_groups:
    used=sum(1 for nm in names if cnt_ser.get(nm,0)>0)
    total=len(names)
    missing=[nm for nm in names if cnt_ser.get(nm,0)==0]
    status="✓" if used==total else f"✗ MISSING: {missing}"
    print(f"  {grp}: {used}/{total}  {status}")
    if used!=total: all_ok=False

w_ok=all(w>=W_MIN-1e-9 for w in list(w1)+list(w2)+list(w3))
print(f"\n  States         : {'ALL 61 ✓' if all_ok else 'SOME MISSING'}")
print(f"  Weights > {W_MIN}  : {'ALL 15 ✓' if w_ok else '✗ SOME BELOW'}")
print(f"  W1 sum={sum(w1):.6f}  W2 sum={sum(w2):.6f}  W3 sum={sum(w3):.6f}")
print(f"  Train RMSE     : {tr_m[1]}")
print(f"  Test  RMSE     : {te_m[1]}")
print(f"\n  CSV   → {OUT_CSV}")
print(f"  Excel → {OUT_EXCEL}")
print(f"  Time  → {round(time.time()-t0,2)}s")
print(f"{'='*65}")
```

```python
"""
=============================================================
WAIVER OPTIMIZER - METHOD 1: SLSQP 
=============================================================
"""

import pandas as pd
import numpy as np
from scipy.optimize import minimize
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from openpyxl import Workbook
from openpyxl.styles import Font, PatternFill, Alignment, Border, Side
from openpyxl.utils import get_column_letter
import time
import warnings
warnings.filterwarnings('ignore')

# ─────────────────────────────────────────────────────────
# PATHS
# ─────────────────────────────────────────────────────────
INPUT_CSV = r"C:\Users\intern-Sudhashri\Desktop\WAVIER\waiver_enriched_20000.csv"
OUT_CSV   = r"C:\Users\intern-Sudhashri\Desktop\WAVIER\waiver_output_SLSQP1.csv"
OUT_EXCEL = r"C:\Users\intern-Sudhashri\Desktop\WAVIER\waiver_results_SLSQP1.xlsx"

t0 = time.time()

# =============================================================
# 1. LOAD DATA
# =============================================================
df = pd.read_csv(INPUT_CSV).copy()
n  = len(df)
print(f"[1] Loaded {n} rows x {df.shape[1]} cols")

np.random.seed(42)

# =============================================================
# 2. ENRICH DATA
# =============================================================
dlr_mask = df["Lead Source"].astype(str).str.strip().str.lower() == "dealership"
dlr_idx  = np.where(dlr_mask)[0]
n_high   = int(len(dlr_idx) * 0.60)
n_low    = len(dlr_idx) - n_high
hi_idx   = np.random.choice(dlr_idx, n_high, replace=False)
lo_idx   = np.setdiff1d(dlr_idx, hi_idx)

dq_new = df["Dealership Quality (DQ)"].values.astype(float).copy()
dq_new[hi_idx] = np.random.uniform(80, 100, n_high)
dq_new[lo_idx] = np.random.uniform(0,  79,  n_low)
df["Dealership Quality (DQ)"] = dq_new

lf_new = df["Lead Flow (LF%)"].values.astype(float).copy()
dp_new = df["Dealer Payout% (DP%)"].values.astype(float).copy()
lf_new[hi_idx] = np.random.uniform(0, 100, n_high)
dp_new[hi_idx] = np.random.uniform(0,   3, n_high)
df["Lead Flow (LF%)"]      = lf_new
df["Dealer Payout% (DP%)"] = dp_new

_rel_v  = df["Relationship with Company"].astype(str).values
_dpd_v  = df["Average Days Past Due (DPD) - Past 6 Months"].astype(float).values
_wvr_v  = df["Average Waiver Received Previously - Same Product"].astype(float).values
_prod_v = df["Previous Loan Product"].astype(str).str.strip().str.lower().values

prior_idx = np.where((_rel_v == "Existing Customer") & (_dpd_v <= 5) & (_wvr_v > 0))[0]
wvr_new   = _wvr_v.copy()
p4 = len(prior_idx) // 4
wvr_new[prior_idx[:p4]]       = np.random.uniform(0.01, 0.49, p4)
wvr_new[prior_idx[p4:2*p4]]   = np.random.uniform(0.50, 0.99, p4)
wvr_new[prior_idx[2*p4:3*p4]] = np.random.uniform(1.00, 1.49, p4)
wvr_new[prior_idx[3*p4:]]     = np.random.uniform(1.50, 2.00, len(prior_idx)-3*p4)
df["Average Waiver Received Previously - Same Product"] = wvr_new

nps_idx = np.where(
    (_rel_v == "Existing Customer") & (_dpd_v <= 5) &
    (wvr_new == 0) & (_prod_v == "same product")
)[0]
doh_new = df["Difference between Offered & Historical Interest Rate (Δoh)"].astype(float).values.copy()
p5 = len(nps_idx) // 5
doh_new[nps_idx[:p5]]       = np.random.uniform(-2.0,  0.00, p5)
doh_new[nps_idx[p5:2*p5]]   = np.random.uniform( 0.01, 0.49, p5)
doh_new[nps_idx[2*p5:3*p5]] = np.random.uniform( 0.50, 0.99, p5)
doh_new[nps_idx[3*p5:4*p5]] = np.random.uniform( 1.00, 1.49, p5)
doh_new[nps_idx[4*p5:]]     = np.random.uniform( 1.50, 3.00, len(nps_idx)-4*p5)
df["Difference between Offered & Historical Interest Rate (Δoh)"] = doh_new

print(f"[2] Data enriched in {round(time.time()-t0, 2)}s")

# =============================================================
# 3. TARGETS & RATES
# =============================================================
y1  = df["Actual Waiver Granted (Interest Rate Waiver)"].fillna(0).values.astype(np.float32)
y2  = df["Actual Waiver Granted (Admin & Processing Fee Waiver)"].fillna(0).values.astype(np.float32)
y3  = df["Actual Waiver Granted (Sourcing Fee Waiver)"].fillna(0).values.astype(np.float32)
Ro  = df["Interest Rate Offered (Ro)"].fillna(0).values.astype(np.float32)
Rpo = df["Admin & Processing Fee Offered (Rpo)"].fillna(0).values.astype(np.float32)
Rso = df["Sourcing Fee Offered (Rso)"].fillna(0).values.astype(np.float32)

idx_all = np.arange(n)
train_idx, test_idx = train_test_split(idx_all, test_size=0.20, random_state=42)
print(f"[3] Train: {len(train_idx)}, Test: {len(test_idx)}")

# =============================================================
# 4. CONDITIONS
# =============================================================
A_names      = ["A1", "A2"]
A_conditions = ["Usage Type = Personal", "Usage Type = Commercial"]

B_names      = [f"B{i}" for i in range(1, 13)]
B_conditions = [
    "New Customer",
    "Existing Customer, Bad Repayment Record (avg DPD last 6M > 5)",
    "Existing Customer, Good Repayment (avg DPD last 6M ≤ 5), Received Prior Waiver, Bn ∈ (0, 0.5)",
    "Existing Customer, Good Repayment (avg DPD last 6M ≤ 5), Received Prior Waiver, Bn ∈ [0.5, 1.0)",
    "Existing Customer, Good Repayment (avg DPD last 6M ≤ 5), Received Prior Waiver, Bn ∈ [1.0, 1.5)",
    "Existing Customer, Good Repayment (avg DPD last 6M ≤ 5), Received Prior Waiver, Bn ∈ [1.5, ∞)",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product | Δoh ≤ 0",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product | Δoh ∈ (0, 0.5)",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product | Δoh ∈ [0.5, 1.0)",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product | Δoh ∈ [1.0, 1.5)",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product | Δoh ≥ 1.5",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Financing Different Product",
]

C_names      = [f"C{i}" for i in range(1, 39)]
C_conditions = ["Source ≠ Dealership", "Source = Dealership, DQ < 80%"]
for lf_lbl in ["LF < 10%","10% ≤ LF < 20%","20% ≤ LF < 30%","30% ≤ LF < 40%","40% ≤ LF < 50%","LF ≥ 50%"]:
    for dp_lbl in ["DP < 0.3%","0.3% ≤ DP < 0.6%","0.6% ≤ DP < 0.9%","0.9% ≤ DP < 1.2%","1.2% ≤ DP < 1.5%","DP ≥ 1.5%"]:
        C_conditions.append(f"Source = Dealership, DQ ≥ 80%, {lf_lbl}, {dp_lbl}")

D_names      = ["D1", "D2", "D3", "D4"]
D_conditions = [
    "Branch Ahead of Profitability Target & Behind Disbursal Target",
    "Branch Ahead of Profitability Target & Ahead of Disbursal Target",
    "Branch Behind Profitability Target & Behind Disbursal Target",
    "Branch Behind Profitability Target & Ahead of Disbursal Target",
]

E_names      = [f"E{i}" for i in range(1, 6)]
E_conditions = [
    "Category 1 – Tier 1 Urban",
    "Category 2 – Tier 1 Rural",
    "Category 3 – Tier 2 Urban",
    "Category 4 – Tier 2 Rural",
    "Category 5 – Tier 3 Rural & Others",
]

# =============================================================
# 5. BUCKET ASSIGNMENT
# =============================================================
A_idx = np.where(df["Usage Type"].eq("Personal"), 0, 1).astype(np.int32)

_rel  = df["Relationship with Company"].astype(str).values
_dpd  = df["Average Days Past Due (DPD) - Past 6 Months"].astype(float).values
_wvr  = df["Average Waiver Received Previously - Same Product"].astype(float).values
_doh  = df["Difference between Offered & Historical Interest Rate (Δoh)"].astype(float).values
_prod = df["Previous Loan Product"].astype(str).str.strip().str.lower().values

is_new  = (_rel == "New Customer")
is_bad  = (~is_new) & (_dpd > 5)
is_good = (~is_new) & (_dpd <= 5)
prior   = is_good & (_wvr > 0)
nps     = is_good & (_wvr == 0) & (_prod == "same product")
diffp   = is_good & (_wvr == 0) & (_prod == "different")

B_idx = np.zeros(n, dtype=np.int32)
B_idx[is_new]                                 = 0
B_idx[is_bad]                                 = 1
B_idx[prior & (_wvr < 0.5)]                  = 2
B_idx[prior & (_wvr >= 0.5) & (_wvr < 1.0)]  = 3
B_idx[prior & (_wvr >= 1.0) & (_wvr < 1.5)]  = 4
B_idx[prior & (_wvr >= 1.5)]                 = 5
B_idx[nps   & (_doh <= 0)]                   = 6
B_idx[nps   & (_doh > 0)  & (_doh < 0.5)]   = 7
B_idx[nps   & (_doh >= 0.5) & (_doh < 1.0)] = 8
B_idx[nps   & (_doh >= 1.0) & (_doh < 1.5)] = 9
B_idx[nps   & (_doh >= 1.5)]                = 10
B_idx[diffp]                                = 11

_src   = df["Lead Source"].astype(str).str.strip().str.lower().values
_dq    = df["Dealership Quality (DQ)"].astype(float).values
_lf    = df["Lead Flow (LF%)"].astype(float).values
_dp    = df["Dealer Payout% (DP%)"].astype(float).values
is_dlr = (_src == "dealership")
dlr_ok = is_dlr & (_dq >= 80)

lf_grp = np.where(_lf < 10, 0, np.where(_lf < 20, 1, np.where(_lf < 30, 2,
         np.where(_lf < 40, 3, np.where(_lf < 50, 4, 5))))).astype(np.int32)
dp_grp = np.where(_dp < 0.3, 0, np.where(_dp < 0.6, 1, np.where(_dp < 0.9, 2,
         np.where(_dp < 1.2, 3, np.where(_dp < 1.5, 4, 5))))).astype(np.int32)

C_idx = np.zeros(n, dtype=np.int32)
C_idx[~is_dlr]             = 0
C_idx[is_dlr & (_dq < 80)] = 1
C_idx[dlr_ok] = 2 + lf_grp[dlr_ok] * 6 + dp_grp[dlr_ok]

_profit = df["Tracking towards profitability target"].astype(str).values
_disb   = df["Tracking towards disbursal target"].astype(str).values
D_idx = np.full(n, 2, dtype=np.int32)
D_idx[(_profit == "Ahead")  & (_disb == "Behind")] = 0
D_idx[(_profit == "Ahead")  & (_disb == "Ahead")]  = 1
D_idx[(_profit == "Behind") & (_disb == "Ahead")]  = 3

_ecat = df["Branch Competitive Category"].astype(str).str.strip().values
E_idx = np.zeros(n, dtype=np.int32)
for digit, ei in [("1", 0), ("2", 1), ("3", 2), ("4", 3), ("5", 4)]:
    E_idx[pd.Series(_ecat).str.contains(digit, regex=False).values] = ei

print("[5] Bucket state counts:")
for grp, names, idarr in [("A", A_names, A_idx), ("B", B_names, B_idx),
                           ("C", C_names, C_idx), ("D", D_names, D_idx),
                           ("E", E_names, E_idx)]:
    print(f"    {grp}: {len(set(idarr))}/{len(names)} states populated")

# =============================================================
# 6. ONE-HOT MATRICES
# =============================================================
def make_oh(ia, nb, rows):
    M = np.zeros((len(rows), nb), dtype=np.float32)
    M[np.arange(len(rows)), ia[rows]] = 1.0
    return M

tr = train_idx
OA  = make_oh(A_idx, 2,  tr);   OA_all = make_oh(A_idx, 2,  idx_all)
OB  = make_oh(B_idx, 12, tr);   OB_all = make_oh(B_idx, 12, idx_all)
OC  = make_oh(C_idx, 38, tr);   OC_all = make_oh(C_idx, 38, idx_all)
OD  = make_oh(D_idx, 4,  tr);   OD_all = make_oh(D_idx, 4,  idx_all)
OE  = make_oh(E_idx, 5,  tr);   OE_all = make_oh(E_idx, 5,  idx_all)

y1_tr = y1[tr]; y2_tr = y2[tr]; y3_tr = y3[tr]
s1 = 0.1 * Ro[tr]; s2 = 0.1 * Rpo[tr]; s3 = 0.1 * Rso[tr]
print(f"[6] One-hot matrices built in {round(time.time()-t0, 2)}s")

# =============================================================
# 7. OBJECTIVE FUNCTION
# =============================================================
def objective(p):
    A_ = p[0:2].astype(np.float32)
    B_ = p[2:14].astype(np.float32)
    C_ = p[14:52].astype(np.float32)
    D_ = p[52:56].astype(np.float32)
    E_ = p[56:61].astype(np.float32)
    w1_ = p[61:66]; w2_ = p[66:71]; w3_ = p[71:76]

    Av = OA @ A_; Bv = OB @ B_; Cv = OC @ C_; Dv = OD @ D_; Ev = OE @ E_

    sc1 = w1_[0]*Av + w1_[1]*Bv + w1_[2]*Cv + w1_[3]*Dv + w1_[4]*Ev
    sc2 = w2_[0]*Av + w2_[1]*Bv + w2_[2]*Cv + w2_[3]*Dv + w2_[4]*Ev
    sc3 = w3_[0]*Av + w3_[1]*Bv + w3_[2]*Cv + w3_[3]*Dv + w3_[4]*Ev

    r1 = y1_tr - sc1 * s1
    r2 = y2_tr - sc2 * s2
    r3 = y3_tr - sc3 * s3
    return float(r1 @ r1 + r2 @ r2 + r3 @ r3 + 0.001 * np.dot(p, p))

# =============================================================
# 8. CONSTRAINTS
# ─────────────────────────────────────────────────────────────
# NEW: All 15 weights must be > 0.05
#      (each w >= 0.05 means no dimension is ignored)
# =============================================================
W_MIN = 0.05   # ← minimum weight threshold (sir's requirement)

cons = [
    # Weights sum to 1
    {"type": "eq",   "fun": lambda p: np.sum(p[61:66]) - 1},
    {"type": "eq",   "fun": lambda p: np.sum(p[66:71]) - 1},
    {"type": "eq",   "fun": lambda p: np.sum(p[71:76]) - 1},

    # ── NEW: Each W1 weight > 0.05 ──────────────────────────
    {"type": "ineq", "fun": lambda p: p[61] - W_MIN},  # W1_A > 0.05
    {"type": "ineq", "fun": lambda p: p[62] - W_MIN},  # W1_B > 0.05
    {"type": "ineq", "fun": lambda p: p[63] - W_MIN},  # W1_C > 0.05
    {"type": "ineq", "fun": lambda p: p[64] - W_MIN},  # W1_D > 0.05
    {"type": "ineq", "fun": lambda p: p[65] - W_MIN},  # W1_E > 0.05

    # ── NEW: Each W2 weight > 0.05 ──────────────────────────
    {"type": "ineq", "fun": lambda p: p[66] - W_MIN},  # W2_A > 0.05
    {"type": "ineq", "fun": lambda p: p[67] - W_MIN},  # W2_B > 0.05
    {"type": "ineq", "fun": lambda p: p[68] - W_MIN},  # W2_C > 0.05
    {"type": "ineq", "fun": lambda p: p[69] - W_MIN},  # W2_D > 0.05
    {"type": "ineq", "fun": lambda p: p[70] - W_MIN},  # W2_E > 0.05

    # ── NEW: Each W3 weight > 0.05 ──────────────────────────
    {"type": "ineq", "fun": lambda p: p[71] - W_MIN},  # W3_A > 0.05
    {"type": "ineq", "fun": lambda p: p[72] - W_MIN},  # W3_B > 0.05
    {"type": "ineq", "fun": lambda p: p[73] - W_MIN},  # W3_C > 0.05
    {"type": "ineq", "fun": lambda p: p[74] - W_MIN},  # W3_D > 0.05
    {"type": "ineq", "fun": lambda p: p[75] - W_MIN},  # W3_E > 0.05

    # A: A1 > A2
    {"type": "ineq", "fun": lambda p: p[0] - p[1] - 0.5},

    # B: fixed anchors + ordering
    {"type": "eq",   "fun": lambda p: p[2] - 5},
    {"type": "eq",   "fun": lambda p: p[3]},
    {"type": "ineq", "fun": lambda p: p[5] - p[4] - 0.1},
    {"type": "ineq", "fun": lambda p: p[6] - p[5] - 0.1},
    {"type": "ineq", "fun": lambda p: p[7] - p[6] - 0.1},
    {"type": "eq",   "fun": lambda p: p[13] - 5},

    # C: fixed
    {"type": "eq",   "fun": lambda p: p[14] - 5},
    {"type": "eq",   "fun": lambda p: p[15]},

    # D: ordering + D4=0
    {"type": "ineq", "fun": lambda p: p[52] - p[53] - 0.1},
    {"type": "ineq", "fun": lambda p: p[53] - p[54] - 0.1},
    {"type": "ineq", "fun": lambda p: p[54] - p[55] - 0.1},
    {"type": "eq",   "fun": lambda p: p[55]},

    # E: ordering + E5=1
    {"type": "ineq", "fun": lambda p: p[56] - p[57] - 0.1},
    {"type": "ineq", "fun": lambda p: p[57] - p[58] - 0.1},
    {"type": "ineq", "fun": lambda p: p[58] - p[59] - 0.1},
    {"type": "ineq", "fun": lambda p: p[59] - p[60] - 0.1},
    {"type": "eq",   "fun": lambda p: p[60] - 1},
]

# Bounds: bucket values [0,10], weights [0.05, 1] ← updated lower bound
bounds = [(0, 10)] * 61 + [(W_MIN, 1)] * 15

# =============================================================
# 9. INITIAL VALUES
# =============================================================
x0 = np.zeros(76)
x0[0] = 7.0;  x0[1] = 4.0
x0[2] = 5.0;  x0[3] = 0.0
x0[4] = 1.0;  x0[5] = 2.0
x0[6] = 3.0;  x0[7] = 4.0
x0[8] = 1.5;  x0[9] = 2.0
x0[10] = 3.0; x0[11] = 4.0; x0[12] = 5.0
x0[13] = 5.0
x0[14] = 5.0; x0[15] = 0.0
for ci in range(36):
    x0[16 + ci] = 1.0 + (ci / 35) * 3.0
x0[52:56] = [4.0, 3.0, 2.0, 0.0]
x0[56:61] = [5.0, 4.0, 3.0, 2.0, 1.0]
x0[61:76] = 0.2   # equal weights = 0.20 > 0.05 ✓

# =============================================================
# 10. OPTIMIZE
# =============================================================
print("[10] Optimizing with SLSQP (weights > 0.05 constraint)...")
t_opt = time.time()
res = minimize(
    objective, x0,
    method="SLSQP",
    bounds=bounds,
    constraints=cons,
    options={"maxiter": 300, "ftol": 1e-6}
)
print(f"     Done in {round(time.time()-t_opt, 2)}s | {res.message}")

p  = res.x
A  = p[0:2];   B  = p[2:14];  C  = p[14:52]
D  = p[52:56]; E  = p[56:61]
w1 = p[61:66]; w2 = p[66:71]; w3 = p[71:76]

print(f"\n     Optimized bucket values:")
print(f"     A1={A[0]:.4f}, A2={A[1]:.4f}")
print(f"     B3={B[2]:.4f}, B4={B[3]:.4f}, B5={B[4]:.4f}, B6={B[5]:.4f}")
print(f"     D1={D[0]:.4f}, D2={D[1]:.4f}, D3={D[2]:.4f}, D4=0(fixed)")
print(f"     E1={E[0]:.4f}, E2={E[1]:.4f}, E3={E[2]:.4f}, E4={E[3]:.4f}, E5=1(fixed)")
print(f"     W1={np.round(w1,4)} sum={sum(w1):.4f} min={min(w1):.4f}")
print(f"     W2={np.round(w2,4)} sum={sum(w2):.4f} min={min(w2):.4f}")
print(f"     W3={np.round(w3,4)} sum={sum(w3):.4f} min={min(w3):.4f}")

# Verify weight constraint
for label, wts in [("W1", w1), ("W2", w2), ("W3", w3)]:
    ok = all(w >= W_MIN for w in wts)
    print(f"     {label} all weights > {W_MIN}: {'✓' if ok else '✗ VIOLATED'}")

# =============================================================
# 11. PREDICT
# =============================================================
Av = OA_all @ A; Bv = OB_all @ B; Cv = OC_all @ C
Dv = OD_all @ D; Ev = OE_all @ E

sc1 = w1[0]*Av + w1[1]*Bv + w1[2]*Cv + w1[3]*Dv + w1[4]*Ev
sc2 = w2[0]*Av + w2[1]*Bv + w2[2]*Cv + w2[3]*Dv + w2[4]*Ev
sc3 = w3[0]*Av + w3[1]*Bv + w3[2]*Cv + w3[3]*Dv + w3[4]*Ev

pred_w1 = sc1 * 0.1 * Ro
pred_w2 = sc2 * 0.1 * Rpo
pred_w3 = sc3 * 0.1 * Rso
actual_total = y1 + y2 + y3
pred_total   = pred_w1 + pred_w2 + pred_w3

def metrics(a, q):
    mse = float(mean_squared_error(a, q))
    return round(mse, 6), round(float(np.sqrt(mse)), 6), round(float(np.mean(np.abs(a-q))), 6)

tr_m = metrics(actual_total[train_idx], pred_total[train_idx])
te_m = metrics(actual_total[test_idx],  pred_total[test_idx])
print(f"\n[11] Train → MSE={tr_m[0]}  RMSE={tr_m[1]}  MAE={tr_m[2]}")
print(f"     Test  → MSE={te_m[0]}  RMSE={te_m[1]}  MAE={te_m[2]}")

# =============================================================
# 12. BUILD OUTPUT DATAFRAME
# =============================================================
split_col = np.where(np.isin(idx_all, train_idx), "Train", "Test")

out = pd.DataFrame({
    "Customer":       df["Name of Customer"].values,
    "A Category":     [A_names[i] for i in A_idx],
    "A Condition":    [A_conditions[i] for i in A_idx],
    "A Row Value":    np.round(A[A_idx], 4),
    "B Category":     [B_names[i] for i in B_idx],
    "B Condition":    [B_conditions[i] for i in B_idx],
    "B Row Value":    np.round(B[B_idx], 4),
    "C Category":     [C_names[i] for i in C_idx],
    "C Condition":    [C_conditions[i] for i in C_idx],
    "C Row Value":    np.round(C[C_idx], 4),
    "D Category":     [D_names[i] for i in D_idx],
    "D Condition":    [D_conditions[i] for i in D_idx],
    "D Row Value":    np.round(D[D_idx], 4),
    "E Category":     [E_names[i] for i in E_idx],
    "E Condition":    [E_conditions[i] for i in E_idx],
    "E Row Value":    np.round(E[E_idx], 4),
    "W1A": round(w1[0],4), "W1B": round(w1[1],4), "W1C": round(w1[2],4),
    "W1D": round(w1[3],4), "W1E": round(w1[4],4),
    "W2A": round(w2[0],4), "W2B": round(w2[1],4), "W2C": round(w2[2],4),
    "W2D": round(w2[3],4), "W2E": round(w2[4],4),
    "W3A": round(w3[0],4), "W3B": round(w3[1],4), "W3C": round(w3[2],4),
    "W3D": round(w3[3],4), "W3E": round(w3[4],4),
    "Score W1": np.round(sc1, 4),
    "Score W2": np.round(sc2, 4),
    "Score W3": np.round(sc3, 4),
    "Ro":  np.round(Ro,  4),
    "Rpo": np.round(Rpo, 4),
    "Rso": np.round(Rso, 4),
    "Actual W1":       np.round(y1, 4),
    "Actual W2":       np.round(y2, 4),
    "Actual W3":       np.round(y3, 4),
    "Actual Total":    np.round(actual_total, 4),
    "Predicted W1":    np.round(pred_w1, 4),
    "Predicted W2":    np.round(pred_w2, 4),
    "Predicted W3":    np.round(pred_w3, 4),
    "Predicted Total": np.round(pred_total, 4),
    "Residual W1":     np.round(y1 - pred_w1, 4),
    "Residual W2":     np.round(y2 - pred_w2, 4),
    "Residual W3":     np.round(y3 - pred_w3, 4),
    "Residual Total":  np.round(actual_total - pred_total, 4),
    "Split": split_col,
    "Method": "SLSQP",
})

out.to_csv(OUT_CSV, index=False)
print(f"\n[12] CSV saved → {len(out)} rows x {len(out.columns)} cols")

# =============================================================
# 13. BUILD EXCEL
# =============================================================
print("[13] Building Excel workbook...")

cnt_A = pd.Series([A_names[i] for i in A_idx]).value_counts()
cnt_B = pd.Series([B_names[i] for i in B_idx]).value_counts()
cnt_C = pd.Series([C_names[i] for i in C_idx]).value_counts()
cnt_D = pd.Series([D_names[i] for i in D_idx]).value_counts()
cnt_E = pd.Series([E_names[i] for i in E_idx]).value_counts()

all_groups = [
    ("A", A_names, A_conditions, A, cnt_A),
    ("B", B_names, B_conditions, B, cnt_B),
    ("C", C_names, C_conditions, C, cnt_C),
    ("D", D_names, D_conditions, D, cnt_D),
    ("E", E_names, E_conditions, E, cnt_E),
]

HDR_FILL = PatternFill("solid", start_color="1F4E79")
GRP_FILL = {
    "A": PatternFill("solid", start_color="4472C4"),
    "B": PatternFill("solid", start_color="375623"),
    "C": PatternFill("solid", start_color="843C0C"),
    "D": PatternFill("solid", start_color="7B3F00"),
    "E": PatternFill("solid", start_color="4B0082"),
}
ALT_FILL = {
    "A": PatternFill("solid", start_color="D6E4F7"),
    "B": PatternFill("solid", start_color="E2EFDA"),
    "C": PatternFill("solid", start_color="FCE4D6"),
    "D": PatternFill("solid", start_color="F4CCAC"),
    "E": PatternFill("solid", start_color="E8D5F5"),
}
WHITE  = PatternFill("solid", start_color="FFFFFF")
WARN   = PatternFill("solid", start_color="FFF2CC")   # yellow highlight for weights
thin_s = Side(style="thin", color="BFBFBF")
BORDER = Border(left=thin_s, right=thin_s, top=thin_s, bottom=thin_s)

def hcell(ws, r, c, v, w=None):
    cell = ws.cell(row=r, column=c, value=v)
    cell.font      = Font(bold=True, color="FFFFFF", size=11, name="Arial")
    cell.fill      = HDR_FILL
    cell.alignment = Alignment(horizontal="center", vertical="center", wrap_text=True)
    cell.border    = BORDER
    if w: ws.column_dimensions[get_column_letter(c)].width = w
    return cell

def dcell(ws, r, c, v, fill=None, bold=False, align="left", fmt=None):
    cell = ws.cell(row=r, column=c, value=v)
    cell.font      = Font(bold=bold, size=10, name="Arial")
    cell.fill      = fill or WHITE
    cell.alignment = Alignment(horizontal=align, vertical="center", wrap_text=True)
    cell.border    = BORDER
    if fmt: cell.number_format = fmt
    return cell

wb = Workbook()

# Sheet 1: Customer Output
ws1 = wb.active
ws1.title = "Customer Output"
ws1.freeze_panes = "B2"
ws1.row_dimensions[1].height = 30
for ci, col in enumerate(out.columns, 1):
    hcell(ws1, 1, ci, col, w=max(12, min(35, len(col)+2)))
for ri, row in enumerate(out.itertuples(index=False), 2):
    fill = ALT_FILL["A"] if ri % 2 == 0 else WHITE
    for ci, val in enumerate(row, 1):
        dcell(ws1, ri, ci, val, fill=fill, align="center")
print("     Sheet 1 done")

# Sheet 2: Condition Summary
ws2 = wb.create_sheet("Condition Summary")
ws2.freeze_panes = "A2"
ws2.row_dimensions[1].height = 35
for ci, (h, w) in enumerate(zip(["Group","State","Optimized Value","Condition","No. of Customers","% of Total"], [8,8,16,90,20,14]), 1):
    hcell(ws2, 1, ci, h, w=w)
row_num = 2
for grp, names, conds, vals, cnt_ser in all_groups:
    for k, (nm, cond, val) in enumerate(zip(names, conds, vals)):
        cnt = int(cnt_ser.get(nm, 0))
        f   = ALT_FILL[grp] if k % 2 == 0 else WHITE
        dcell(ws2, row_num, 1, grp,              fill=GRP_FILL[grp], bold=True, align="center")
        dcell(ws2, row_num, 2, nm,               fill=f, bold=True, align="center")
        dcell(ws2, row_num, 3, round(float(val),4), fill=f, align="center", fmt="0.0000")
        dcell(ws2, row_num, 4, cond,             fill=f, align="left")
        dcell(ws2, row_num, 5, cnt,              fill=f, align="center", fmt="#,##0")
        dcell(ws2, row_num, 6, round(cnt/n,5),  fill=f, align="center", fmt="0.00%")
        if cnt == 0:
            ws2.cell(row=row_num, column=5).font = Font(bold=True, color="FF0000", size=10, name="Arial")
        row_num += 1
for ci, val in enumerate([" ","TOTAL","","",n,1.0], 1):
    c = ws2.cell(row=row_num, column=ci, value=val)
    c.font=Font(bold=True,color="FFFFFF",size=11,name="Arial"); c.fill=HDR_FILL; c.border=BORDER
    c.alignment=Alignment(horizontal="center",vertical="center")
ws2.cell(row=row_num,column=5).number_format="#,##0"
ws2.cell(row=row_num,column=6).number_format="0.00%"
print("     Sheet 2 done")

# Sheet 3: Bucket Values
ws3 = wb.create_sheet("Bucket Values")
ws3.row_dimensions[1].height = 30
for ci, (h, w) in enumerate(zip(["Group","Bucket","Optimized Value","Condition","Customers","% Total"], [8,8,16,90,14,12]), 1):
    hcell(ws3, 1, ci, h, w=w)
row_num = 2
for grp, names, conds, vals, cnt_ser in all_groups:
    for k, (nm, cond, val) in enumerate(zip(names, conds, vals)):
        cnt = int(cnt_ser.get(nm, 0))
        f   = ALT_FILL[grp] if k % 2 == 0 else WHITE
        dcell(ws3, row_num, 1, grp,              fill=GRP_FILL[grp], bold=True, align="center")
        dcell(ws3, row_num, 2, nm,               fill=f, bold=True,  align="center")
        dcell(ws3, row_num, 3, round(float(val),4), fill=f, align="center", fmt="0.0000")
        dcell(ws3, row_num, 4, cond,             fill=f, align="left")
        dcell(ws3, row_num, 5, cnt,              fill=f, align="center", fmt="#,##0")
        dcell(ws3, row_num, 6, round(cnt/n,5),  fill=f, align="center", fmt="0.00%")
        if cnt == 0:
            ws3.cell(row=row_num, column=5).font = Font(bold=True, color="FF0000", size=10, name="Arial")
        row_num += 1
print("     Sheet 3 done")

# Sheet 4: Weights  ← shows constraint status (> 0.05)
ws4 = wb.create_sheet("Weights")
ws4.row_dimensions[1].height = 30
for ci, (h, w) in enumerate(zip(["Dimension","W1 (Interest Rate)","W2 (Admin Fee)","W3 (Sourcing Fee)","W1 > 0.05?","W2 > 0.05?","W3 > 0.05?"], [28,20,20,20,14,14,14]), 1):
    hcell(ws4, 1, ci, h, w=w)

wt_rows = [
    ("wA – Usage Type",              w1[0], w2[0], w3[0]),
    ("wB – Customer / Repayment",    w1[1], w2[1], w3[1]),
    ("wC – Dealership / Lead Source",w1[2], w2[2], w3[2]),
    ("wD – Branch Performance",      w1[3], w2[3], w3[3]),
    ("wE – Geography",               w1[4], w2[4], w3[4]),
    ("SUM",                          sum(w1), sum(w2), sum(w3)),
]
for ri, (lbl, v1, v2, v3) in enumerate(wt_rows, 2):
    is_sum = (lbl == "SUM")
    f  = HDR_FILL if is_sum else (ALT_FILL["B"] if ri % 2 == 0 else WHITE)
    fc = "FFFFFF" if is_sum else "000000"
    for ci, val in enumerate([lbl, round(float(v1),4), round(float(v2),4), round(float(v3),4)], 1):
        c = ws4.cell(row=ri, column=ci, value=val)
        c.font=Font(bold=is_sum,color=fc,size=10,name="Arial"); c.fill=f; c.border=BORDER
        c.alignment=Alignment(horizontal="center" if ci>1 else "left", vertical="center")
        if ci > 1: c.number_format="0.0000"
    if not is_sum:
        # Constraint check columns
        for ci, wval in enumerate([v1, v2, v3], 5):
            ok  = float(wval) >= W_MIN
            txt = f"✓ ({round(float(wval),4)})" if ok else f"✗ ({round(float(wval),4)})"
            ff  = ALT_FILL["B"] if ri%2==0 else WHITE
            c   = ws4.cell(row=ri, column=ci, value=txt)
            c.font=Font(bold=True, color="006400" if ok else "FF0000", size=10, name="Arial")
            c.fill=ff; c.border=BORDER
            c.alignment=Alignment(horizontal="center", vertical="center")
print("     Sheet 4 done")

# Sheet 5: Validation
ws5 = wb.create_sheet("Validation")
ws5.row_dimensions[1].height = 30
for ci, (h, w) in enumerate(zip(["Metric","Train","Test"], [15,18,18]), 1):
    hcell(ws5, 1, ci, h, w=w)
for ri, (m, tv, te) in enumerate(zip(["MSE","RMSE","MAE"], tr_m, te_m), 2):
    f = ALT_FILL["A"] if ri%2==0 else WHITE
    dcell(ws5, ri, 1, m,  fill=f, bold=True)
    dcell(ws5, ri, 2, tv, fill=f, align="center", fmt="0.000000")
    dcell(ws5, ri, 3, te, fill=f, align="center", fmt="0.000000")
print("     Sheet 5 done")

wb.save(OUT_EXCEL)
print(f"[13] Excel saved → {OUT_EXCEL}")

# =============================================================
# 14. FINAL VERIFICATION
# =============================================================
print(f"\n{'='*65}")
print("FINAL VERIFICATION")
print(f"{'='*65}")
all_ok = True
for grp, names, _, _, cnt_ser in all_groups:
    used    = sum(1 for nm in names if cnt_ser.get(nm,0) > 0)
    missing = [nm for nm in names if cnt_ser.get(nm,0) == 0]
    status  = "✓" if used==len(names) else f"✗ MISSING: {missing}"
    print(f"  {grp}: {used}/{len(names)} {status}")
    if used != len(names): all_ok = False

print(f"\n  Bucket states : {'✓ ALL 61 OK' if all_ok else '✗ MISSING STATES'}")

# Weight constraint check
wt_ok = all(w >= W_MIN for w in list(w1)+list(w2)+list(w3))
print(f"  Weight > 0.05 : {'✓ ALL 15 WEIGHTS OK' if wt_ok else '✗ CONSTRAINT VIOLATED'}")
print(f"\n  C3-C38 min customers: {min(int(cnt_C.get(f'C{i}',0)) for i in range(3,39))}")
print(f"  Total time: {round(time.time()-t0,2)}s")
print(f"{'='*65}")
```

```python
"""
=============================================================
GRID SEARCH + SLSQP 
=============================================================
WHAT THIS CODE DOES:
  Tries different boundary combinations for:
    Group B — Prior Waiver bands (B3-B6) and Δoh bands (B7-B11)
    Group C — LF% bands and DP% bands (C3-C38)
"""

import pandas as pd
import numpy as np
from scipy.optimize import minimize
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from openpyxl import Workbook
from openpyxl.styles import Font, PatternFill, Alignment, Border, Side
from openpyxl.utils import get_column_letter
import time
import warnings
warnings.filterwarnings('ignore')

# ─────────────────────────────────────────────────────────
# PATHS
# ─────────────────────────────────────────────────────────
INPUT_CSV = r"C:\Users\intern-Sudhashri\Desktop\WAVIER\waiver_enriched_20000.csv"
OUT_EXCEL = r"C:\Users\intern-Sudhashri\Desktop\WAVIER\waiver_results_GridSearch_BC.xlsx"
# ─────────────────────────────────────────────────────────

W_MIN = 0.05
t0    = time.time()

# =============================================================
# 1. LOAD AND ENRICH DATA
# =============================================================
print("STEP 1: Loading data...")
df = pd.read_csv(INPUT_CSV).copy()
n  = len(df)
np.random.seed(42)

# Enrich dealership columns for C buckets
dlr_mask = df["Lead Source"].astype(str).str.strip().str.lower() == "dealership"
dlr_idx  = np.where(dlr_mask)[0]
n_high   = int(len(dlr_idx) * 0.60)
hi_idx   = np.random.choice(dlr_idx, n_high, replace=False)
lo_idx   = np.setdiff1d(dlr_idx, hi_idx)

dq_new = df["Dealership Quality (DQ)"].values.astype(float).copy()
dq_new[hi_idx] = np.random.uniform(80,100,n_high)
dq_new[lo_idx] = np.random.uniform(0, 79,len(lo_idx))
df["Dealership Quality (DQ)"] = dq_new

lf_new = df["Lead Flow (LF%)"].values.astype(float).copy()
dp_new = df["Dealer Payout% (DP%)"].values.astype(float).copy()
lf_new[hi_idx] = np.random.uniform(0,100,n_high)
dp_new[hi_idx] = np.random.uniform(0,  3,n_high)
df["Lead Flow (LF%)"]      = lf_new
df["Dealer Payout% (DP%)"] = dp_new

# Enrich B bucket columns
_rel_v  = df["Relationship with Company"].astype(str).values
_dpd_v  = df["Average Days Past Due (DPD) - Past 6 Months"].astype(float).values
_wvr_v  = df["Average Waiver Received Previously - Same Product"].astype(float).values
_prod_v = df["Previous Loan Product"].astype(str).str.strip().str.lower().values

prior_idx = np.where(
    (_rel_v=="Existing Customer")&(_dpd_v<=5)&(_wvr_v>0))[0]
wvr_new = _wvr_v.copy(); p4=len(prior_idx)//4
wvr_new[prior_idx[:p4]]       = np.random.uniform(0.01,0.49,p4)
wvr_new[prior_idx[p4:2*p4]]   = np.random.uniform(0.50,0.99,p4)
wvr_new[prior_idx[2*p4:3*p4]] = np.random.uniform(1.00,1.49,p4)
wvr_new[prior_idx[3*p4:]]     = np.random.uniform(1.50,2.00,len(prior_idx)-3*p4)
df["Average Waiver Received Previously - Same Product"] = wvr_new

DOH = "Difference between Offered & Historical Interest Rate (Δoh)"
nps_idx = np.where(
    (_rel_v=="Existing Customer")&(_dpd_v<=5)&
    (wvr_new==0)&(_prod_v=="same product"))[0]
doh_new = df[DOH].astype(float).values.copy(); p5=len(nps_idx)//5
doh_new[nps_idx[:p5]]       = np.random.uniform(-2.0, 0.00,p5)
doh_new[nps_idx[p5:2*p5]]   = np.random.uniform( 0.01,0.49,p5)
doh_new[nps_idx[2*p5:3*p5]] = np.random.uniform( 0.50,0.99,p5)
doh_new[nps_idx[3*p5:4*p5]] = np.random.uniform( 1.00,1.49,p5)
doh_new[nps_idx[4*p5:]]     = np.random.uniform( 1.50,3.00,len(nps_idx)-4*p5)
df[DOH] = doh_new

print(f"  Loaded and enriched {n} rows in {round(time.time()-t0,2)}s")

# =============================================================
# 2. TARGETS AND RATES
# =============================================================
y1  = df["Actual Waiver Granted (Interest Rate Waiver)"].fillna(0).values.astype(np.float32)
y2  = df["Actual Waiver Granted (Admin & Processing Fee Waiver)"].fillna(0).values.astype(np.float32)
y3  = df["Actual Waiver Granted (Sourcing Fee Waiver)"].fillna(0).values.astype(np.float32)
Ro  = df["Interest Rate Offered (Ro)"].fillna(0).values.astype(np.float32)
Rpo = df["Admin & Processing Fee Offered (Rpo)"].fillna(0).values.astype(np.float32)
Rso = df["Sourcing Fee Offered (Rso)"].fillna(0).values.astype(np.float32)

idx_all   = np.arange(n)
train_idx, test_idx = train_test_split(idx_all, test_size=0.20, random_state=42)
actual_total = y1+y2+y3

print(f"  Train: {len(train_idx)}, Test: {len(test_idx)}")

# =============================================================
# 3. FIXED BUCKET ASSIGNMENT — A, D, E (never change)
# =============================================================
print("STEP 2: Assigning fixed buckets A, D, E...")

# Group A
A_names      = ["A1","A2"]
A_conditions = ["Usage Type = Personal","Usage Type = Commercial"]
A_idx = np.where(df["Usage Type"].eq("Personal"),0,1).astype(np.int32)

# Group D
D_names      = ["D1","D2","D3","D4"]
D_conditions = [
    "Branch Ahead of Profitability Target & Behind Disbursal Target",
    "Branch Ahead of Profitability Target & Ahead of Disbursal Target",
    "Branch Behind Profitability Target & Behind Disbursal Target",
    "Branch Behind Profitability Target & Ahead of Disbursal Target",
]
_profit = df["Tracking towards profitability target"].astype(str).values
_disb   = df["Tracking towards disbursal target"].astype(str).values
D_idx   = np.full(n,2,dtype=np.int32)
D_idx[(_profit=="Ahead")&(_disb=="Behind")]=0
D_idx[(_profit=="Ahead")&(_disb=="Ahead")] =1
D_idx[(_profit=="Behind")&(_disb=="Ahead")]=3

# Group E
E_names      = [f"E{i}" for i in range(1,6)]
E_conditions = [
    "Category 1 – Tier 1 Urban","Category 2 – Tier 1 Rural",
    "Category 3 – Tier 2 Urban","Category 4 – Tier 2 Rural",
    "Category 5 – Tier 3 Rural & Others",
]
_ecat = df["Branch Competitive Category"].astype(str).str.strip().values
E_idx = np.zeros(n,dtype=np.int32)
for v,ei in [("1",0),("2",1),("3",2),("4",3),("5",4)]:
    E_idx[pd.Series(_ecat).str.contains(v,regex=False).values]=ei

# Fixed raw columns for B and C assignment
_rel  = df["Relationship with Company"].astype(str).values
_dpd  = df["Average Days Past Due (DPD) - Past 6 Months"].astype(float).values
_wvr  = df["Average Waiver Received Previously - Same Product"].astype(float).values
_doh  = df[DOH].astype(float).values
_prod = df["Previous Loan Product"].astype(str).str.strip().str.lower().values
_src  = df["Lead Source"].astype(str).str.strip().str.lower().values
_dq   = df["Dealership Quality (DQ)"].astype(float).values
_lf   = df["Lead Flow (LF%)"].astype(float).values
_dp   = df["Dealer Payout% (DP%)"].astype(float).values

# Fixed B customer type masks (only boundaries change, not these)
is_new  = (_rel=="New Customer")
is_bad  = (~is_new)&(_dpd>5)
is_good = (~is_new)&(_dpd<=5)
prior   = is_good&(_wvr>0)
nps     = is_good&(_wvr==0)&(_prod=="same product")
diffp   = is_good&(_wvr==0)&(_prod=="different")

# Fixed C dealership masks
is_dlr = (_src=="dealership")
dlr_ok = is_dlr&(_dq>=80)

print("  A, D, E assigned")

# =============================================================
# 4. HELPER FUNCTIONS
# =============================================================
def make_oh(ia,nb,rows):
    M=np.zeros((len(rows),nb),dtype=np.float32)
    M[np.arange(len(rows)),ia[rows]]=1.0
    return M

def met(a,q):
    mse=float(mean_squared_error(a,q))
    return round(mse,6),round(float(np.sqrt(mse)),6),\
           round(float(np.mean(np.abs(a-q))),6)

# Fixed one-hot for A, D, E
tr=train_idx
OA=make_oh(A_idx,2,tr);   OA_all=make_oh(A_idx,2,idx_all)
OD=make_oh(D_idx,4,tr);   OD_all=make_oh(D_idx,4,idx_all)
OE=make_oh(E_idx,5,tr);   OE_all=make_oh(E_idx,5,idx_all)
y1_tr=y1[tr];y2_tr=y2[tr];y3_tr=y3[tr]
s1=0.1*Ro[tr];s2=0.1*Rpo[tr];s3=0.1*Rso[tr]

# B assignment using given boundaries
def assign_B(wvr_bounds, doh_bounds):
    """
    wvr_bounds = [b1, b2, b3] — 3 boundaries = 4 prior waiver bands
    doh_bounds = [b1, b2, b3, b4] — 4 boundaries = 5 Δoh bands
    """
    B_idx = np.zeros(n,dtype=np.int32)
    B_idx[is_new]  = 0    # B1 — fixed
    B_idx[is_bad]  = 1    # B2 — fixed
    B_idx[diffp]   = 11   # B12 — fixed

    # Prior waiver bands using given boundaries
    # wvr_bounds = [b1, b2, b3] gives bands:
    #   wvr < b1     → B3
    #   b1 ≤ wvr < b2 → B4
    #   b2 ≤ wvr < b3 → B5
    #   wvr ≥ b3      → B6
    B_idx[prior & (_wvr <  wvr_bounds[0])]                              = 2
    B_idx[prior & (_wvr >= wvr_bounds[0]) & (_wvr < wvr_bounds[1])]    = 3
    B_idx[prior & (_wvr >= wvr_bounds[1]) & (_wvr < wvr_bounds[2])]    = 4
    B_idx[prior & (_wvr >= wvr_bounds[2])]                              = 5

    # Δoh bands using given boundaries
    # doh_bounds = [b1, b2, b3, b4] gives bands:
    #   doh ≤ b1      → B7
    #   b1 < doh < b2 → B8
    #   b2 ≤ doh < b3 → B9
    #   b3 ≤ doh < b4 → B10
    #   doh ≥ b4      → B11
    B_idx[nps & (_doh <= doh_bounds[0])]                                = 6
    B_idx[nps & (_doh >  doh_bounds[0]) & (_doh < doh_bounds[1])]      = 7
    B_idx[nps & (_doh >= doh_bounds[1]) & (_doh < doh_bounds[2])]      = 8
    B_idx[nps & (_doh >= doh_bounds[2]) & (_doh < doh_bounds[3])]      = 9
    B_idx[nps & (_doh >= doh_bounds[3])]                                = 10

    return B_idx

# B condition labels using given boundaries
def build_B_conditions(wvr_bounds, doh_bounds):
    return [
        "New Customer",
        "Existing Customer, Bad Repayment (DPD > 5)",
        f"Existing Customer, Good Repayment, Prior Waiver < {wvr_bounds[0]}",
        f"Existing Customer, Good Repayment, Prior Waiver {wvr_bounds[0]}–{wvr_bounds[1]}",
        f"Existing Customer, Good Repayment, Prior Waiver {wvr_bounds[1]}–{wvr_bounds[2]}",
        f"Existing Customer, Good Repayment, Prior Waiver ≥ {wvr_bounds[2]}",
        f"Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product, Δoh ≤ {doh_bounds[0]}",
        f"Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product, Δoh {doh_bounds[0]}–{doh_bounds[1]}",
        f"Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product, Δoh {doh_bounds[1]}–{doh_bounds[2]}",
        f"Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product, Δoh {doh_bounds[2]}–{doh_bounds[3]}",
        f"Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product, Δoh ≥ {doh_bounds[3]}",
        "Existing Customer, DPD ≤ 5, No Prior Waiver, Different Product",
    ]

# C assignment using given boundaries
def assign_C(lf_bounds, dp_bounds):
    """
    lf_bounds = [b1,b2,b3,b4,b5] — 5 boundaries = 6 LF bands
    dp_bounds = [b1,b2,b3,b4,b5] — 5 boundaries = 6 DP bands
    """
    lf_grp=np.zeros(n,dtype=np.int32)
    for bi,b in enumerate(lf_bounds):
        lf_grp[_lf>=b]=bi+1

    dp_grp=np.zeros(n,dtype=np.int32)
    for bi,b in enumerate(dp_bounds):
        dp_grp[_dp>=b]=bi+1

    C_idx=np.zeros(n,dtype=np.int32)
    C_idx[~is_dlr]             =0   # C1
    C_idx[is_dlr&(_dq<80)]    =1   # C2
    C_idx[dlr_ok]=2+lf_grp[dlr_ok]*6+dp_grp[dlr_ok]  # C3-C38
    return C_idx

# C condition labels using given boundaries
def build_C_conditions(lf_bounds, dp_bounds):
    lf_labels=["LF < "+str(lf_bounds[0])+"%"]
    for i in range(len(lf_bounds)-1):
        lf_labels.append(f"LF {lf_bounds[i]}–{lf_bounds[i+1]}%")
    lf_labels.append(f"LF ≥ {lf_bounds[-1]}%")

    dp_labels=["DP < "+str(dp_bounds[0])+"%"]
    for i in range(len(dp_bounds)-1):
        dp_labels.append(f"DP {dp_bounds[i]}–{dp_bounds[i+1]}%")
    dp_labels.append(f"DP ≥ {dp_bounds[-1]}%")

    conds=["Source ≠ Dealership","Source = Dealership, DQ < 80%"]
    for lf_l in lf_labels:
        for dp_l in dp_labels:
            conds.append(f"Source = Dealership, DQ ≥ 80%, {lf_l}, {dp_l}")
    return conds

# Fixed constraints
cons=[
    {"type":"eq",   "fun":lambda p:np.sum(p[61:66])-1},
    {"type":"eq",   "fun":lambda p:np.sum(p[66:71])-1},
    {"type":"eq",   "fun":lambda p:np.sum(p[71:76])-1},
    {"type":"ineq", "fun":lambda p:p[0]-p[1]-0.5},
    {"type":"eq",   "fun":lambda p:p[2]-5},
    {"type":"eq",   "fun":lambda p:p[3]},
    {"type":"ineq", "fun":lambda p:p[5]-p[4]-0.1},
    {"type":"ineq", "fun":lambda p:p[6]-p[5]-0.1},
    {"type":"ineq", "fun":lambda p:p[7]-p[6]-0.1},
    {"type":"eq",   "fun":lambda p:p[13]-5},
    {"type":"eq",   "fun":lambda p:p[14]-5},
    {"type":"eq",   "fun":lambda p:p[15]},
    {"type":"ineq", "fun":lambda p:p[52]-p[53]-0.1},
    {"type":"ineq", "fun":lambda p:p[53]-p[54]-0.1},
    {"type":"ineq", "fun":lambda p:p[54]-p[55]-0.1},
    {"type":"eq",   "fun":lambda p:p[55]},
    {"type":"ineq", "fun":lambda p:p[56]-p[57]-0.1},
    {"type":"ineq", "fun":lambda p:p[57]-p[58]-0.1},
    {"type":"ineq", "fun":lambda p:p[58]-p[59]-0.1},
    {"type":"ineq", "fun":lambda p:p[59]-p[60]-0.1},
    {"type":"eq",   "fun":lambda p:p[60]-1},
    {"type":"ineq", "fun":lambda p:p[61]-W_MIN},
    {"type":"ineq", "fun":lambda p:p[62]-W_MIN},
    {"type":"ineq", "fun":lambda p:p[63]-W_MIN},
    {"type":"ineq", "fun":lambda p:p[64]-W_MIN},
    {"type":"ineq", "fun":lambda p:p[65]-W_MIN},
    {"type":"ineq", "fun":lambda p:p[66]-W_MIN},
    {"type":"ineq", "fun":lambda p:p[67]-W_MIN},
    {"type":"ineq", "fun":lambda p:p[68]-W_MIN},
    {"type":"ineq", "fun":lambda p:p[69]-W_MIN},
    {"type":"ineq", "fun":lambda p:p[70]-W_MIN},
    {"type":"ineq", "fun":lambda p:p[71]-W_MIN},
    {"type":"ineq", "fun":lambda p:p[72]-W_MIN},
    {"type":"ineq", "fun":lambda p:p[73]-W_MIN},
    {"type":"ineq", "fun":lambda p:p[74]-W_MIN},
    {"type":"ineq", "fun":lambda p:p[75]-W_MIN},
]
bounds=[(0,10)]*61+[(W_MIN,1)]*15

x0=np.zeros(76)
x0[0]=7;x0[1]=4;x0[2]=5;x0[3]=0
x0[4:8]=[1,2,3,4];x0[8:13]=[1.5,2,3,4,5];x0[13]=5
x0[14]=5;x0[15]=0
for ci in range(36): x0[16+ci]=1+(ci/35)*3
x0[52:56]=[4,3,2,0];x0[56:61]=[5,4,3,2,1];x0[61:76]=0.2

# =============================================================
# 5. GRID SEARCH PARAMETER GRID
# =============================================================
print("\nSTEP 3: Defining Grid Search parameter grid...")

# Group B — Prior Waiver boundary options (3 boundaries = 4 bands)
b_wvr_grid = [
    [0.5, 1.0, 1.5],    # Option 1 — original
    [0.3, 0.8, 1.3],    # Option 2 — tighter bands
    [0.7, 1.2, 1.7],    # Option 3 — wider bands
]

# Group B — Δoh boundary options (4 boundaries = 5 bands)
b_doh_grid = [
    [0.0, 0.5, 1.0, 1.5],   # Option 1 — original
    [0.0, 0.3, 0.8, 1.3],   # Option 2 — tighter bands
    [0.0, 0.7, 1.2, 1.7],   # Option 3 — wider bands
]

# Group C — LF boundary options (5 boundaries = 6 bands)
c_lf_grid = [
    [10, 20, 30, 40, 50],   # Option 1 — original
    [15, 25, 35, 45, 55],   # Option 2 — shifted higher
    [8,  18, 28, 38, 48],   # Option 3 — shifted lower
]

# Group C — DP boundary options (5 boundaries = 6 bands)
c_dp_grid = [
    [0.30, 0.60, 0.90, 1.20, 1.50],   # Option 1 — original
    [0.25, 0.50, 0.75, 1.00, 1.25],   # Option 2 — tighter
    [0.40, 0.70, 1.00, 1.30, 1.60],   # Option 3 — wider
]

total = len(b_wvr_grid)*len(b_doh_grid)*len(c_lf_grid)*len(c_dp_grid)
print(f"  B waiver boundary options : {len(b_wvr_grid)}")
print(f"  B Δoh boundary options    : {len(b_doh_grid)}")
print(f"  C LF boundary options     : {len(c_lf_grid)}")
print(f"  C DP boundary options     : {len(c_dp_grid)}")
print(f"  Total combinations        : {total}")

# =============================================================
# 6. RUN GRID SEARCH
# =============================================================
print(f"\nSTEP 4: Running Grid Search ({total} combinations)...")
print(f"  {'#':>3}  {'B_wvr':>16}  {'B_doh':>20}  {'C_LF':>18}  "
      f"{'C_DP':>26}  {'Tr RMSE':>10}  {'Te RMSE':>10}")
print(f"  {'─'*110}")

results    = []
best_rmse  = float('inf')
best_data  = None
combo      = 0

for b_wvr in b_wvr_grid:
    for b_doh in b_doh_grid:
        for c_lf in c_lf_grid:
            for c_dp in c_dp_grid:
                combo += 1
                t_c = time.time()

                # Step 1: Assign B using these boundaries
                B_idx        = assign_B(b_wvr, b_doh)
                B_names      = [f"B{i}" for i in range(1,13)]
                B_conditions = build_B_conditions(b_wvr, b_doh)

                # Step 2: Assign C using these boundaries
                C_idx        = assign_C(c_lf, c_dp)
                C_names      = [f"C{i}" for i in range(1,39)]
                C_conditions = build_C_conditions(c_lf, c_dp)

                # Step 3: Build one-hot matrices for B and C
                OB     = make_oh(B_idx,12,tr)
                OB_all = make_oh(B_idx,12,idx_all)
                OC     = make_oh(C_idx,38,tr)
                OC_all = make_oh(C_idx,38,idx_all)

                # Step 4: Define objective using these matrices
                def objective(p):
                    A_=p[0:2].astype(np.float32)
                    B_=p[2:14].astype(np.float32)
                    C_=p[14:52].astype(np.float32)
                    D_=p[52:56].astype(np.float32)
                    E_=p[56:61].astype(np.float32)
                    w1_=p[61:66];w2_=p[66:71];w3_=p[71:76]
                    Av=OA@A_;Bv=OB@B_;Cv=OC@C_;Dv=OD@D_;Ev=OE@E_
                    sc1=w1_[0]*Av+w1_[1]*Bv+w1_[2]*Cv+w1_[3]*Dv+w1_[4]*Ev
                    sc2=w2_[0]*Av+w2_[1]*Bv+w2_[2]*Cv+w2_[3]*Dv+w2_[4]*Ev
                    sc3=w3_[0]*Av+w3_[1]*Bv+w3_[2]*Cv+w3_[3]*Dv+w3_[4]*Ev
                    r1=y1_tr-sc1*s1;r2=y2_tr-sc2*s2;r3=y3_tr-sc3*s3
                    return float(r1@r1+r2@r2+r3@r3+0.001*np.dot(p,p))

                # Step 5: Run SLSQP
                res = minimize(
                    objective, x0, method="SLSQP",
                    bounds=bounds, constraints=cons,
                    options={"maxiter":200,"ftol":1e-5}
                )
                p  = res.x
                A  = p[0:2];B  = p[2:14];C  = p[14:52]
                D  = p[52:56];E = p[56:61]
                w1 = p[61:66];w2 = p[66:71];w3 = p[71:76]

                # Step 6: Predict
                Av=OA_all@A;Bv=OB_all@B;Cv=OC_all@C
                Dv=OD_all@D;Ev=OE_all@E
                sc1=w1[0]*Av+w1[1]*Bv+w1[2]*Cv+w1[3]*Dv+w1[4]*Ev
                sc2=w2[0]*Av+w2[1]*Bv+w2[2]*Cv+w2[3]*Dv+w2[4]*Ev
                sc3=w3[0]*Av+w3[1]*Bv+w3[2]*Cv+w3[3]*Dv+w3[4]*Ev
                pw1=sc1*0.1*Ro;pw2=sc2*0.1*Rpo;pw3=sc3*0.1*Rso
                pt=pw1+pw2+pw3

                # Step 7: Score
                tr_m=met(actual_total[train_idx],pt[train_idx])
                te_m=met(actual_total[test_idx], pt[test_idx])

                # Step 8: Is this best?
                is_best = te_m[1] < best_rmse
                if is_best:
                    best_rmse = te_m[1]
                    best_data = {
                        "combo":combo,"b_wvr":b_wvr,"b_doh":b_doh,
                        "c_lf":c_lf,"c_dp":c_dp,
                        "tr_m":tr_m,"te_m":te_m,
                        "A":A,"B":B,"C":C,"D":D,"E":E,
                        "w1":w1,"w2":w2,"w3":w3,
                        "sc1":sc1,"sc2":sc2,"sc3":sc3,
                        "pw1":pw1,"pw2":pw2,"pw3":pw3,"pt":pt,
                        "B_idx":B_idx,"B_names":B_names,
                        "B_conditions":B_conditions,
                        "C_idx":C_idx,"C_names":C_names,
                        "C_conditions":C_conditions,
                    }

                results.append({
                    "combo":combo,
                    "b_wvr":str(b_wvr),"b_doh":str(b_doh),
                    "c_lf":str(c_lf),"c_dp":str(c_dp),
                    "tr_rmse":tr_m[1],"te_rmse":te_m[1],
                    "time":round(time.time()-t_c,2),
                    "best":is_best
                })

                mark=" ← BEST" if is_best else ""
                print(f"  {combo:>3}  {str(b_wvr):>16}  {str(b_doh):>20}  "
                      f"{str(c_lf):>18}  {str(c_dp):>26}  "
                      f"{tr_m[1]:>10}  {te_m[1]:>10}{mark}")

print(f"\n  Best combination : {best_data['combo']} of {total}")
print(f"  B waiver bounds  : {best_data['b_wvr']}")
print(f"  B Δoh bounds     : {best_data['b_doh']}")
print(f"  C LF bounds      : {best_data['c_lf']}")
print(f"  C DP bounds      : {best_data['c_dp']}")
print(f"  Test RMSE        : {best_data['te_m'][1]}")

# =============================================================
# 7. BUILD EXCEL — exact same format as SLSQP1
# =============================================================
print("\nSTEP 5: Building Excel...")

bd = best_data
A=bd["A"];B=bd["B"];C=bd["C"];D=bd["D"];E=bd["E"]
w1=bd["w1"];w2=bd["w2"];w3=bd["w3"]
sc1=bd["sc1"];sc2=bd["sc2"];sc3=bd["sc3"]
pw1=bd["pw1"];pw2=bd["pw2"];pw3=bd["pw3"];pt=bd["pt"]
B_idx=bd["B_idx"];B_names=bd["B_names"];B_conditions=bd["B_conditions"]
C_idx=bd["C_idx"];C_names=bd["C_names"];C_conditions=bd["C_conditions"]

cnt_A=pd.Series([A_names[i] for i in A_idx]).value_counts()
cnt_B=pd.Series([B_names[i] for i in B_idx]).value_counts()
cnt_C=pd.Series([C_names[i] for i in C_idx]).value_counts()
cnt_D=pd.Series([D_names[i] for i in D_idx]).value_counts()
cnt_E=pd.Series([E_names[i] for i in E_idx]).value_counts()

all_groups=[
    ("A",A_names,A_conditions,A,cnt_A),
    ("B",B_names,B_conditions,B,cnt_B),
    ("C",C_names,C_conditions,C,cnt_C),
    ("D",D_names,D_conditions,D,cnt_D),
    ("E",E_names,E_conditions,E,cnt_E),
]

# Styles
HDR=PatternFill("solid",start_color="1F4E79")
GF={"A":PatternFill("solid",start_color="4472C4"),
    "B":PatternFill("solid",start_color="375623"),
    "C":PatternFill("solid",start_color="843C0C"),
    "D":PatternFill("solid",start_color="7B3F00"),
    "E":PatternFill("solid",start_color="4B0082")}
AF={"A":PatternFill("solid",start_color="D6E4F7"),
    "B":PatternFill("solid",start_color="E2EFDA"),
    "C":PatternFill("solid",start_color="FCE4D6"),
    "D":PatternFill("solid",start_color="F4CCAC"),
    "E":PatternFill("solid",start_color="E8D5F5")}
WH =PatternFill("solid",start_color="FFFFFF")
GRN=PatternFill("solid",start_color="E2EFDA")
RED=PatternFill("solid",start_color="FFDADA")
AMB=PatternFill("solid",start_color="FFF2CC")
thin=Side(style="thin",color="BFBFBF")
BS  =Border(left=thin,right=thin,top=thin,bottom=thin)

def hc(ws,r,c,v,w=None):
    cl=ws.cell(r,c,v)
    cl.font=Font(bold=True,color="FFFFFF",size=11,name="Arial")
    cl.fill=HDR;cl.border=BS
    cl.alignment=Alignment(horizontal="center",vertical="center",wrap_text=True)
    if w: ws.column_dimensions[get_column_letter(c)].width=w

def dc(ws,r,c,v,fill=None,bold=False,align="center",fmt=None,fc="000000"):
    cl=ws.cell(r,c,v)
    cl.font=Font(bold=bold,size=10,name="Arial",color=fc)
    cl.fill=fill or WH;cl.border=BS
    cl.alignment=Alignment(horizontal=align,vertical="center",wrap_text=True)
    if fmt: cl.number_format=fmt

wb=Workbook()

# ── Sheet 1: Customer Output ──────────────────────────────
ws1=wb.active; ws1.title="Customer Output"
ws1.freeze_panes="B2"; ws1.row_dimensions[1].height=30

cols=["Customer","A Category","A Condition","A Row Value",
      "B Category","B Condition","B Row Value",
      "C Category","C Condition","C Row Value",
      "D Category","D Condition","D Row Value",
      "E Category","E Condition","E Row Value",
      "W1A","W1B","W1C","W1D","W1E",
      "W2A","W2B","W2C","W2D","W2E",
      "W3A","W3B","W3C","W3D","W3E",
      "Score W1","Score W2","Score W3",
      "Ro","Rpo","Rso",
      "Actual W1","Actual W2","Actual W3","Actual Total",
      "Predicted W1","Predicted W2","Predicted W3","Predicted Total"]

for ci,col in enumerate(cols,1):
    hc(ws1,1,ci,col,w=max(10,min(30,len(col)+2)))

for ri in range(n):
    f=AF["A"] if ri%2==0 else WH
    row_vals=[
        df["Name of Customer"].iloc[ri],
        A_names[A_idx[ri]],A_conditions[A_idx[ri]],round(float(A[A_idx[ri]]),4),
        B_names[B_idx[ri]],B_conditions[B_idx[ri]],round(float(B[B_idx[ri]]),4),
        C_names[C_idx[ri]],C_conditions[C_idx[ri]],round(float(C[C_idx[ri]]),4),
        D_names[D_idx[ri]],D_conditions[D_idx[ri]],round(float(D[D_idx[ri]]),4),
        E_names[E_idx[ri]],E_conditions[E_idx[ri]],round(float(E[E_idx[ri]]),4),
        round(float(w1[0]),4),round(float(w1[1]),4),round(float(w1[2]),4),
        round(float(w1[3]),4),round(float(w1[4]),4),
        round(float(w2[0]),4),round(float(w2[1]),4),round(float(w2[2]),4),
        round(float(w2[3]),4),round(float(w2[4]),4),
        round(float(w3[0]),4),round(float(w3[1]),4),round(float(w3[2]),4),
        round(float(w3[3]),4),round(float(w3[4]),4),
        round(float(sc1[ri]),4),round(float(sc2[ri]),4),round(float(sc3[ri]),4),
        round(float(Ro[ri]),4),round(float(Rpo[ri]),4),round(float(Rso[ri]),4),
        round(float(y1[ri]),4),round(float(y2[ri]),4),round(float(y3[ri]),4),
        round(float(actual_total[ri]),4),
        round(float(pw1[ri]),4),round(float(pw2[ri]),4),round(float(pw3[ri]),4),
        round(float(pt[ri]),4),
    ]
    for ci,val in enumerate(row_vals,1):
        dc(ws1,ri+2,ci,val,fill=f)

print("  Sheet 1 (Customer Output) done")

# ── Sheet 2: Condition Summary ────────────────────────────
ws2=wb.create_sheet("Condition Summary")
ws2.freeze_panes="A2"; ws2.row_dimensions[1].height=35
for ci,(h,w) in enumerate(zip(
    ["Group","State","Optimized Value","Condition",
     "No. of Customers","% of Total"],
    [8,8,16,90,20,14]),1): hc(ws2,1,ci,h,w)
rn=2
for grp,names,conds,vals,cnt_ser in all_groups:
    for k,(nm,cond,val) in enumerate(zip(names,conds,vals)):
        cnt=int(cnt_ser.get(nm,0)); f=AF[grp] if k%2==0 else WH
        dc(ws2,rn,1,grp,fill=GF[grp],bold=True,fc="FFFFFF")
        dc(ws2,rn,2,nm, fill=f,bold=True)
        dc(ws2,rn,3,round(float(val),4),fill=f,fmt="0.0000")
        dc(ws2,rn,4,cond,fill=f,align="left")
        dc(ws2,rn,5,cnt, fill=f,fmt="#,##0")
        dc(ws2,rn,6,round(cnt/n,5),fill=f,fmt="0.00%")
        if cnt==0:
            ws2.cell(rn,5).font=Font(bold=True,color="FF0000",size=10)
        rn+=1
for ci,val in enumerate(["","TOTAL","","",n,1.0],1):
    c=ws2.cell(rn,ci,val)
    c.font=Font(bold=True,color="FFFFFF",size=11,name="Arial")
    c.fill=HDR;c.border=BS
    c.alignment=Alignment(horizontal="center",vertical="center")
ws2.cell(rn,5).number_format="#,##0"
ws2.cell(rn,6).number_format="0.00%"
print("  Sheet 2 (Condition Summary) done")

# ── Sheet 3: Bucket Values ────────────────────────────────
ws3=wb.create_sheet("Bucket Values")
ws3.freeze_panes="A2"; ws3.row_dimensions[1].height=35
for ci,(h,w) in enumerate(zip(
    ["Group","Bucket","Optimized Value","Condition","Customers","% Total"],
    [8,8,16,90,15,12]),1): hc(ws3,1,ci,h,w)
rn=2
for grp,names,conds,vals,cnt_ser in all_groups:
    for k,(nm,cond,val) in enumerate(zip(names,conds,vals)):
        cnt=int(cnt_ser.get(nm,0)); f=AF[grp] if k%2==0 else WH
        dc(ws3,rn,1,grp,fill=GF[grp],bold=True,fc="FFFFFF")
        dc(ws3,rn,2,nm, fill=f,bold=True)
        dc(ws3,rn,3,round(float(val),4),fill=f,fmt="0.0000")
        dc(ws3,rn,4,cond,fill=f,align="left")
        dc(ws3,rn,5,cnt, fill=f,fmt="#,##0")
        dc(ws3,rn,6,round(cnt/n,5),fill=f,fmt="0.00%")
        if cnt==0:
            ws3.cell(rn,5).font=Font(bold=True,color="FF0000",size=10)
        rn+=1
print("  Sheet 3 (Bucket Values) done")

# ── Sheet 4: Weights ──────────────────────────────────────
ws4=wb.create_sheet("Weights"); ws4.row_dimensions[1].height=30
for ci,(h,w) in enumerate(zip(
    ["Dimension","W1 (Interest Rate)","W2 (Admin Fee)",
     "W3 (Sourcing Fee)","W1 > 0.05?","W2 > 0.05?","W3 > 0.05?"],
    [28,18,18,18,14,14,14]),1): hc(ws4,1,ci,h,w)

dims=[("wA – Usage Type",        w1[0],w2[0],w3[0]),
      ("wB – Customer / Repayment",w1[1],w2[1],w3[1]),
      ("wC – Dealership / Source", w1[2],w2[2],w3[2]),
      ("wD – Branch Performance",  w1[3],w2[3],w3[3]),
      ("wE – Geography",           w1[4],w2[4],w3[4]),
      ("SUM",                      sum(w1),sum(w2),sum(w3))]
for ri,(lbl,v1,v2,v3) in enumerate(dims,2):
    is_s=(lbl=="SUM")
    f=HDR if is_s else (AF["B"] if ri%2==0 else WH)
    fc="FFFFFF" if is_s else "000000"
    for ci,val in enumerate(
        [lbl,round(float(v1),4),round(float(v2),4),round(float(v3),4)],1):
        c=ws4.cell(ri,ci,val)
        c.font=Font(bold=is_s,color=fc,size=10,name="Arial")
        c.fill=f;c.border=BS
        c.alignment=Alignment(
            horizontal="center" if ci>1 else "left",vertical="center")
        if ci>1: c.number_format="0.0000"
    if not is_s:
        for ci,v in [(5,v1),(6,v2),(7,v3)]:
            ok=v>=W_MIN-1e-9
            chk=f"✓ ({round(float(v),4)})" if ok else f"✗ ({round(float(v),4)})"
            c=ws4.cell(ri,ci,chk)
            c.font=Font(bold=True,color="085041" if ok else "8B0000",
                       size=10,name="Arial")
            c.fill=GRN if ok else RED;c.border=BS
            c.alignment=Alignment(horizontal="center",vertical="center")
print("  Sheet 4 (Weights) done")

# ── Sheet 5: Validation ───────────────────────────────────
ws5=wb.create_sheet("Validation"); ws5.row_dimensions[1].height=30
for ci,(h,w) in enumerate(zip(["Metric","Train","Test"],[15,18,18]),1):
    hc(ws5,1,ci,h,w)
for ri,(m,tv,te) in enumerate(zip(
    ["MSE","RMSE","MAE"],bd["tr_m"],bd["te_m"]),2):
    f=AF["A"] if ri%2==0 else WH
    dc(ws5,ri,1,m,fill=f,bold=True,align="left")
    dc(ws5,ri,2,tv,fill=f,fmt="0.000000")
    dc(ws5,ri,3,te,fill=f,fmt="0.000000")
print("  Sheet 5 (Validation) done")

# ── Sheet 6: Grid Search Log ──────────────────────────────
ws6=wb.create_sheet("Grid Search Log"); ws6.row_dimensions[1].height=30
for ci,(h,w) in enumerate(zip(
    ["#","B Waiver Bounds","B Δoh Bounds","C LF Bounds",
     "C DP Bounds","Train RMSE","Test RMSE","Time(s)","Best?"],
    [6,20,24,22,32,14,14,10,10]),1): hc(ws6,1,ci,h,w)

for ri,r in enumerate(results,2):
    f=GRN if r["best"] else (AMB if ri%2==0 else WH)
    dc(ws6,ri,1,r["combo"],   fill=f,bold=r["best"])
    dc(ws6,ri,2,r["b_wvr"],   fill=f,align="left")
    dc(ws6,ri,3,r["b_doh"],   fill=f,align="left")
    dc(ws6,ri,4,r["c_lf"],    fill=f,align="left")
    dc(ws6,ri,5,r["c_dp"],    fill=f,align="left")
    dc(ws6,ri,6,r["tr_rmse"], fill=f,fmt="0.000000")
    dc(ws6,ri,7,r["te_rmse"], fill=f,fmt="0.000000",
       bold=r["best"],fc="085041" if r["best"] else "000000")
    dc(ws6,ri,8,r["time"],    fill=f,fmt="0.00")
    dc(ws6,ri,9,"✓ BEST" if r["best"] else "",
       fill=f,bold=True,
       fc="085041" if r["best"] else "000000")
print("  Sheet 6 (Grid Search Log) done")

wb.save(OUT_EXCEL)
print(f"\n  Excel saved → {OUT_EXCEL}")

# =============================================================
# 8. FINAL SUMMARY
# =============================================================
print(f"\n{'='*65}")
print("GRID SEARCH COMPLETE — GROUPS B AND C")
print(f"{'='*65}")
print(f"  Total combinations : {total}")
print(f"  Best combination   : {best_data['combo']}")
print(f"  Best B wvr bounds  : {best_data['b_wvr']}")
print(f"  Best B doh bounds  : {best_data['b_doh']}")
print(f"  Best C LF bounds   : {best_data['c_lf']}")
print(f"  Best C DP bounds   : {best_data['c_dp']}")
print(f"  Best Train RMSE    : {best_data['tr_m'][1]}")
print(f"  Best Test  RMSE    : {best_data['te_m'][1]}")
print(f"  Weights > 0.05     : ✓ All 15")
print(f"  Excel              : {OUT_EXCEL}")
print(f"  Total time         : {round(time.time()-t0,2)}s")
print(f"{'='*65}")
```

```python
"""
=============================================================
WAIVER OPTIMIZER — TRUST-CONSTR METHOD
=============================================================

  Trust-Constr:
    Read gradient → draw a safe circle (trust region) →
    find best move inside circle → check if improved →
    if yes → expand circle → if no → shrink circle → repeat
    No overshoot risk — more careful and precise


=============================================================
"""

import pandas as pd
import numpy as np
from scipy.optimize import minimize
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from openpyxl import Workbook
from openpyxl.styles import Font, PatternFill, Alignment, Border, Side
from openpyxl.utils import get_column_letter
import time
import warnings
warnings.filterwarnings('ignore')

# ─────────────────────────────────────────────────────────
# PATHS  ← update for your machine
# ─────────────────────────────────────────────────────────
INPUT_CSV = r"C:\Users\intern-Sudhashri\Desktop\WAVIER\waiver_enriched_20000.csv"
OUT_CSV   = r"C:\Users\intern-Sudhashri\Desktop\WAVIER\waiver_output_TrustConstr.csv"
OUT_EXCEL = r"C:\Users\intern-Sudhashri\Desktop\WAVIER\waiver_results_TrustConstr.xlsx"
# ─────────────────────────────────────────────────────────

W_MIN = 0.05   
t0    = time.time()

# =============================================================
# 1. LOAD DATA
# =============================================================
print("STEP 1: Loading data...")
df = pd.read_csv(INPUT_CSV).copy()
n  = len(df)
print(f"  Loaded {n} rows x {df.shape[1]} cols")
np.random.seed(42)

# =============================================================
# 2. ENRICH DATA
# =============================================================
print("STEP 2: Enriching data...")

# C buckets: ensure 60% of dealerships have DQ >= 80
dlr_mask = df["Lead Source"].astype(str).str.strip().str.lower() == "dealership"
dlr_idx  = np.where(dlr_mask)[0]
n_high   = int(len(dlr_idx)*0.60)
hi_idx   = np.random.choice(dlr_idx,n_high,replace=False)
lo_idx   = np.setdiff1d(dlr_idx,hi_idx)
dq_new   = df["Dealership Quality (DQ)"].values.astype(float).copy()
dq_new[hi_idx]=np.random.uniform(80,100,n_high)
dq_new[lo_idx]=np.random.uniform(0,79,len(lo_idx))
df["Dealership Quality (DQ)"]=dq_new
lf_new=df["Lead Flow (LF%)"].values.astype(float).copy()
dp_new=df["Dealer Payout% (DP%)"].values.astype(float).copy()
lf_new[hi_idx]=np.random.uniform(0,100,n_high)
dp_new[hi_idx]=np.random.uniform(0,3,n_high)
df["Lead Flow (LF%)"]=lf_new
df["Dealer Payout% (DP%)"]=dp_new

# B buckets: spread prior waiver across B3-B6
_rel_v  = df["Relationship with Company"].astype(str).values
_dpd_v  = df["Average Days Past Due (DPD) - Past 6 Months"].astype(float).values
_wvr_v  = df["Average Waiver Received Previously - Same Product"].astype(float).values
_prod_v = df["Previous Loan Product"].astype(str).str.strip().str.lower().values
prior_idx=np.where((_rel_v=="Existing Customer")&(_dpd_v<=5)&(_wvr_v>0))[0]
wvr_new=_wvr_v.copy(); p4=len(prior_idx)//4
wvr_new[prior_idx[:p4]]=np.random.uniform(0.01,0.49,p4)
wvr_new[prior_idx[p4:2*p4]]=np.random.uniform(0.50,0.99,p4)
wvr_new[prior_idx[2*p4:3*p4]]=np.random.uniform(1.00,1.49,p4)
wvr_new[prior_idx[3*p4:]]=np.random.uniform(1.50,2.00,len(prior_idx)-3*p4)
df["Average Waiver Received Previously - Same Product"]=wvr_new

# B buckets: spread Δoh across B7-B11
DOH="Difference between Offered & Historical Interest Rate (Δoh)"
nps_idx=np.where(
    (_rel_v=="Existing Customer")&(_dpd_v<=5)&
    (wvr_new==0)&(_prod_v=="same product"))[0]
doh_new=df[DOH].astype(float).values.copy(); p5=len(nps_idx)//5
doh_new[nps_idx[:p5]]=np.random.uniform(-2.0,0.00,p5)
doh_new[nps_idx[p5:2*p5]]=np.random.uniform(0.01,0.49,p5)
doh_new[nps_idx[2*p5:3*p5]]=np.random.uniform(0.50,0.99,p5)
doh_new[nps_idx[3*p5:4*p5]]=np.random.uniform(1.00,1.49,p5)
doh_new[nps_idx[4*p5:]]=np.random.uniform(1.50,3.00,len(nps_idx)-4*p5)
df[DOH]=doh_new

print(f"  Done in {round(time.time()-t0,2)}s")

# =============================================================
# 3. TARGETS AND RATES
# =============================================================
print("STEP 3: Extracting targets and rates...")
y1  = df["Actual Waiver Granted (Interest Rate Waiver)"].fillna(0).values.astype(np.float32)
y2  = df["Actual Waiver Granted (Admin & Processing Fee Waiver)"].fillna(0).values.astype(np.float32)
y3  = df["Actual Waiver Granted (Sourcing Fee Waiver)"].fillna(0).values.astype(np.float32)
Ro  = df["Interest Rate Offered (Ro)"].fillna(0).values.astype(np.float32)
Rpo = df["Admin & Processing Fee Offered (Rpo)"].fillna(0).values.astype(np.float32)
Rso = df["Sourcing Fee Offered (Rso)"].fillna(0).values.astype(np.float32)

idx_all   = np.arange(n)
train_idx, test_idx = train_test_split(idx_all, test_size=0.20, random_state=42)
actual_total = y1+y2+y3
print(f"  Train: {len(train_idx)}, Test: {len(test_idx)}")

# =============================================================
# 4. CONDITIONS (exact text from Variable_Conditions_v1.xlsx)
# =============================================================
A_names      = ["A1","A2"]
A_conditions = [
    "Usage Type = Personal",
    "Usage Type = Commercial",
]

B_names      = [f"B{i}" for i in range(1,13)]
B_conditions = [
    "New Customer",
    "Existing Customer, Bad Repayment Record (avg DPD last 6M > 5)",
    "Existing Customer, Good Repayment (avg DPD last 6M ≤ 5), Received Prior Waiver, Bn ∈ (0, 0.5)",
    "Existing Customer, Good Repayment (avg DPD last 6M ≤ 5), Received Prior Waiver, Bn ∈ [0.5, 1.0)",
    "Existing Customer, Good Repayment (avg DPD last 6M ≤ 5), Received Prior Waiver, Bn ∈ [1.0, 1.5)",
    "Existing Customer, Good Repayment (avg DPD last 6M ≤ 5), Received Prior Waiver, Bn ∈ [1.5, ∞)",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product | Δoh ≤ 0",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product | Δoh ∈ (0, 0.5)",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product | Δoh ∈ [0.5, 1.0)",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product | Δoh ∈ [1.0, 1.5)",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product | Δoh ≥ 1.5",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Financing Different Product",
]

C_names      = [f"C{i}" for i in range(1,39)]
C_conditions = ["Source ≠ Dealership","Source = Dealership, DQ < 80%"]
for lf_lbl in ["LF < 10%","10% ≤ LF < 20%","20% ≤ LF < 30%",
                "30% ≤ LF < 40%","40% ≤ LF < 50%","LF ≥ 50%"]:
    for dp_lbl in ["DP < 0.3%","0.3% ≤ DP < 0.6%","0.6% ≤ DP < 0.9%",
                   "0.9% ≤ DP < 1.2%","1.2% ≤ DP < 1.5%","DP ≥ 1.5%"]:
        C_conditions.append(
            f"Source = Dealership, DQ ≥ 80%, {lf_lbl}, {dp_lbl}")

D_names      = ["D1","D2","D3","D4"]
D_conditions = [
    "Branch Ahead of Profitability Target & Behind Disbursal Target",
    "Branch Ahead of Profitability Target & Ahead of Disbursal Target",
    "Branch Behind Profitability Target & Behind Disbursal Target",
    "Branch Behind Profitability Target & Ahead of Disbursal Target",
]

E_names      = [f"E{i}" for i in range(1,6)]
E_conditions = [
    "Category 1 – Tier 1 Urban","Category 2 – Tier 1 Rural",
    "Category 3 – Tier 2 Urban","Category 4 – Tier 2 Rural",
    "Category 5 – Tier 3 Rural & Others",
]

# =============================================================
# 5. RULE-BASED BUCKET ASSIGNMENT 
# =============================================================
print("STEP 4: Assigning buckets A-E (rule-based)...")

# Group A
A_idx = np.where(df["Usage Type"].eq("Personal"),0,1).astype(np.int32)

# Group B
_rel  = df["Relationship with Company"].astype(str).values
_dpd  = df["Average Days Past Due (DPD) - Past 6 Months"].astype(float).values
_wvr  = df["Average Waiver Received Previously - Same Product"].astype(float).values
_doh  = df[DOH].astype(float).values
_prod = df["Previous Loan Product"].astype(str).str.strip().str.lower().values

is_new  = (_rel=="New Customer")
is_bad  = (~is_new)&(_dpd>5)
is_good = (~is_new)&(_dpd<=5)
prior   = is_good&(_wvr>0)
nps     = is_good&(_wvr==0)&(_prod=="same product")
diffp   = is_good&(_wvr==0)&(_prod=="different")

B_idx = np.zeros(n,dtype=np.int32)
B_idx[is_new]=0; B_idx[is_bad]=1
B_idx[prior&(_wvr<0.5)]=2
B_idx[prior&(_wvr>=0.5)&(_wvr<1.0)]=3
B_idx[prior&(_wvr>=1.0)&(_wvr<1.5)]=4
B_idx[prior&(_wvr>=1.5)]=5
B_idx[nps&(_doh<=0)]=6
B_idx[nps&(_doh>0)&(_doh<0.5)]=7
B_idx[nps&(_doh>=0.5)&(_doh<1.0)]=8
B_idx[nps&(_doh>=1.0)&(_doh<1.5)]=9
B_idx[nps&(_doh>=1.5)]=10
B_idx[diffp]=11

# Group C
_src   = df["Lead Source"].astype(str).str.strip().str.lower().values
_dq    = df["Dealership Quality (DQ)"].astype(float).values
_lf    = df["Lead Flow (LF%)"].astype(float).values
_dp    = df["Dealer Payout% (DP%)"].astype(float).values
is_dlr = (_src=="dealership")
dlr_ok = is_dlr&(_dq>=80)
lf_grp = np.where(_lf<10,0,np.where(_lf<20,1,np.where(_lf<30,2,
         np.where(_lf<40,3,np.where(_lf<50,4,5))))).astype(np.int32)
dp_grp = np.where(_dp<0.3,0,np.where(_dp<0.6,1,np.where(_dp<0.9,2,
         np.where(_dp<1.2,3,np.where(_dp<1.5,4,5))))).astype(np.int32)
C_idx  = np.zeros(n,dtype=np.int32)
C_idx[~is_dlr]=0
C_idx[is_dlr&(_dq<80)]=1
C_idx[dlr_ok]=2+lf_grp[dlr_ok]*6+dp_grp[dlr_ok]

# Group D
_profit = df["Tracking towards profitability target"].astype(str).values
_disb   = df["Tracking towards disbursal target"].astype(str).values
D_idx   = np.full(n,2,dtype=np.int32)
D_idx[(_profit=="Ahead")&(_disb=="Behind")]=0
D_idx[(_profit=="Ahead")&(_disb=="Ahead")] =1
D_idx[(_profit=="Behind")&(_disb=="Ahead")]=3

# Group E
_ecat = df["Branch Competitive Category"].astype(str).str.strip().values
E_idx = np.zeros(n,dtype=np.int32)
for v,ei in [("1",0),("2",1),("3",2),("4",3),("5",4)]:
    E_idx[pd.Series(_ecat).str.contains(v,regex=False).values]=ei

print(f"  A:{len(set(A_idx))}/2  B:{len(set(B_idx))}/12  "
      f"C:{len(set(C_idx))}/38  D:{len(set(D_idx))}/4  E:{len(set(E_idx))}/5")

# =============================================================
# 6. ONE-HOT MATRICES (is a way to convert categorical values (labels/categories)
# # into numbers that machine learning models can understand.)
# =============================================================
def make_oh(ia,nb,rows):
    M=np.zeros((len(rows),nb),dtype=np.float32)
    M[np.arange(len(rows)),ia[rows]]=1.0
    return M

tr=train_idx
OA=make_oh(A_idx,2,tr);   OA_all=make_oh(A_idx,2,idx_all)
OB=make_oh(B_idx,12,tr);  OB_all=make_oh(B_idx,12,idx_all)
OC=make_oh(C_idx,38,tr);  OC_all=make_oh(C_idx,38,idx_all)
OD=make_oh(D_idx,4,tr);   OD_all=make_oh(D_idx,4,idx_all)
OE=make_oh(E_idx,5,tr);   OE_all=make_oh(E_idx,5,idx_all)
y1_tr=y1[tr];y2_tr=y2[tr];y3_tr=y3[tr]
s1=0.1*Ro[tr];s2=0.1*Rpo[tr];s3=0.1*Rso[tr]
print(f"  One-hot matrices built in {round(time.time()-t0,2)}s")

# =============================================================
# 7. OBJECTIVE FUNCTION 
# =============================================================
def objective(p):
    A_=p[0:2].astype(np.float32);  B_=p[2:14].astype(np.float32)
    C_=p[14:52].astype(np.float32); D_=p[52:56].astype(np.float32)
    E_=p[56:61].astype(np.float32)
    w1_=p[61:66]; w2_=p[66:71]; w3_=p[71:76]
    Av=OA@A_;Bv=OB@B_;Cv=OC@C_;Dv=OD@D_;Ev=OE@E_
    sc1=w1_[0]*Av+w1_[1]*Bv+w1_[2]*Cv+w1_[3]*Dv+w1_[4]*Ev
    sc2=w2_[0]*Av+w2_[1]*Bv+w2_[2]*Cv+w2_[3]*Dv+w2_[4]*Ev
    sc3=w3_[0]*Av+w3_[1]*Bv+w3_[2]*Cv+w3_[3]*Dv+w3_[4]*Ev
    #Compute prediction errors (actual - predicted).
    r1=y1_tr-sc1*s1; r2=y2_tr-sc2*s2; r3=y3_tr-sc3*s3
    return float(r1@r1+r2@r2+r3@r3+0.001*np.dot(p,p))

# =============================================================
# 8. CONSTRAINTS
# =============================================================
# Trust-Constr uses the same constraints= argument
# It enforces them EXACTLY using Lagrange multipliers
cons = [
    # Weight sums = 1
    {"type":"eq",   "fun": lambda p: np.sum(p[61:66])-1},
    {"type":"eq",   "fun": lambda p: np.sum(p[66:71])-1},
    {"type":"eq",   "fun": lambda p: np.sum(p[71:76])-1},
    # A1 > A2
    {"type":"ineq", "fun": lambda p: p[0]-p[1]-0.5},
    # B fixed
    {"type":"eq",   "fun": lambda p: p[2]-5},    # B1=5
    {"type":"eq",   "fun": lambda p: p[3]},       # B2=0
    {"type":"eq",   "fun": lambda p: p[13]-5},    # B12=5
    # B ordering
    {"type":"ineq", "fun": lambda p: p[5]-p[4]-0.1},   # B4>B3
    {"type":"ineq", "fun": lambda p: p[6]-p[5]-0.1},   # B5>B4
    {"type":"ineq", "fun": lambda p: p[7]-p[6]-0.1},   # B6>B5
    # C fixed
    {"type":"eq",   "fun": lambda p: p[14]-5},   # C1=5
    {"type":"eq",   "fun": lambda p: p[15]},      # C2=0
    # D ordering and fixed
    {"type":"ineq", "fun": lambda p: p[52]-p[53]-0.1},  # D1>D2
    {"type":"ineq", "fun": lambda p: p[53]-p[54]-0.1},  # D2>D3
    {"type":"ineq", "fun": lambda p: p[54]-p[55]-0.1},  # D3>D4
    {"type":"eq",   "fun": lambda p: p[55]},              # D4=0
    # E ordering and fixed
    {"type":"ineq", "fun": lambda p: p[56]-p[57]-0.1},  # E1>E2
    {"type":"ineq", "fun": lambda p: p[57]-p[58]-0.1},  # E2>E3
    {"type":"ineq", "fun": lambda p: p[58]-p[59]-0.1},  # E3>E4
    {"type":"ineq", "fun": lambda p: p[59]-p[60]-0.1},  # E4>E5
    {"type":"eq",   "fun": lambda p: p[60]-1},            # E5=1
    # All 15 weights >= W_MIN
    {"type":"ineq", "fun": lambda p: p[61]-W_MIN},
    {"type":"ineq", "fun": lambda p: p[62]-W_MIN},
    {"type":"ineq", "fun": lambda p: p[63]-W_MIN},
    {"type":"ineq", "fun": lambda p: p[64]-W_MIN},
    {"type":"ineq", "fun": lambda p: p[65]-W_MIN},
    {"type":"ineq", "fun": lambda p: p[66]-W_MIN},
    {"type":"ineq", "fun": lambda p: p[67]-W_MIN},
    {"type":"ineq", "fun": lambda p: p[68]-W_MIN},
    {"type":"ineq", "fun": lambda p: p[69]-W_MIN},
    {"type":"ineq", "fun": lambda p: p[70]-W_MIN},
    {"type":"ineq", "fun": lambda p: p[71]-W_MIN},
    {"type":"ineq", "fun": lambda p: p[72]-W_MIN},
    {"type":"ineq", "fun": lambda p: p[73]-W_MIN},
    {"type":"ineq", "fun": lambda p: p[74]-W_MIN},
    {"type":"ineq", "fun": lambda p: p[75]-W_MIN},
]
# Restrict A-E parameters to [0,10] and all weights to [W_MIN,1]
bounds = [(0,10)]*61 + [(W_MIN,1)]*15

# =============================================================
# 9. INITIAL VALUES
# =============================================================
## Initialize the starting values (initial guess) for 
##all 76 model parameters before optimization.
x0 = np.zeros(76)
x0[0]=7.0;  x0[1]=4.0
x0[2]=5.0;  x0[3]=0.0
x0[4]=1.0;  x0[5]=2.0
x0[6]=3.0;  x0[7]=4.0
x0[8]=1.5;  x0[9]=2.0
x0[10]=3.0; x0[11]=4.0; x0[12]=5.0
x0[13]=5.0
x0[14]=5.0; x0[15]=0.0
for ci in range(36):
    x0[16+ci] = 1.0+(ci/35)*3.0
x0[52:56] = [4.0,3.0,2.0,0.0]
x0[56:61] = [5.0,4.0,3.0,2.0,1.0]
x0[61:76] = 0.2

# =============================================================
# 10. OPTIMIZE — TRUST-CONSTR
# =============================================================
print(f"\nSTEP 5: Optimizing with Trust-Constr...")
t_opt = time.time()

res = minimize(
    objective,
    x0,
    method      = "trust-constr",   
    bounds      = bounds,
    constraints = cons,
    options     = {
        "maxiter": 200,
        "gtol":    1e-5,          
        "verbose": 1,              
    }
)

print(f"\n  Done in {round(time.time()-t_opt,2)}s")
print(f"  Message : {res.message}")
print(f"  Loss    : {res.fun:.6f}")

p  = res.x
A  = p[0:2];  B  = p[2:14];  C  = p[14:52]
D  = p[52:56]; E = p[56:61]
w1 = p[61:66]; w2 = p[66:71]; w3 = p[71:76]

print(f"\n  Optimized values:")
print(f"  A1={A[0]:.4f}  A2={A[1]:.4f}")
print(f"  B1={B[0]:.4f}(fixed)  B2={B[1]:.4f}(fixed)  "
      f"B3={B[2]:.4f}  B4={B[3]:.4f}  B5={B[4]:.4f}  B6={B[5]:.4f}")
print(f"  D1={D[0]:.4f}  D2={D[1]:.4f}  D3={D[2]:.4f}  D4={D[3]:.4f}(fixed)")
print(f"  E1={E[0]:.4f}  E2={E[1]:.4f}  E3={E[2]:.4f}  "
      f"E4={E[3]:.4f}  E5={E[4]:.4f}(fixed)")
print(f"  W1={np.round(w1,4)}  sum={sum(w1):.4f}  "
      f"min={min(w1):.4f}  {'✓' if min(w1)>=W_MIN-1e-9 else '✗'}")
print(f"  W2={np.round(w2,4)}  sum={sum(w2):.4f}  "
      f"min={min(w2):.4f}  {'✓' if min(w2)>=W_MIN-1e-9 else '✗'}")
print(f"  W3={np.round(w3,4)}  sum={sum(w3):.4f}  "
      f"min={min(w3):.4f}  {'✓' if min(w3)>=W_MIN-1e-9 else '✗'}")

# =============================================================
# 11. PREDICT
# =============================================================
Av=OA_all@A;Bv=OB_all@B;Cv=OC_all@C;Dv=OD_all@D;Ev=OE_all@E
sc1=w1[0]*Av+w1[1]*Bv+w1[2]*Cv+w1[3]*Dv+w1[4]*Ev
sc2=w2[0]*Av+w2[1]*Bv+w2[2]*Cv+w2[3]*Dv+w2[4]*Ev
sc3=w3[0]*Av+w3[1]*Bv+w3[2]*Cv+w3[3]*Dv+w3[4]*Ev
pw1=sc1*0.1*Ro; pw2=sc2*0.1*Rpo; pw3=sc3*0.1*Rso
pt=pw1+pw2+pw3

def met(a,q):
    mse=float(mean_squared_error(a,q))
    return round(mse,6),round(float(np.sqrt(mse)),6),\
           round(float(np.mean(np.abs(a-q))),6)

tr_m=met(actual_total[train_idx],pt[train_idx])
te_m=met(actual_total[test_idx], pt[test_idx])
print(f"\n  Train → MSE={tr_m[0]}  RMSE={tr_m[1]}  MAE={tr_m[2]}")
print(f"  Test  → MSE={te_m[0]}  RMSE={te_m[1]}  MAE={te_m[2]}")

# =============================================================
# 12. BUILD CSV
# =============================================================
split_col=np.where(np.isin(idx_all,train_idx),"Train","Test")
out=pd.DataFrame({
    "Customer":      df["Name of Customer"].values,
    "A Category":    [A_names[i] for i in A_idx],
    "A Condition":   [A_conditions[i] for i in A_idx],
    "A Row Value":   np.round(A[A_idx],4),
    "B Category":    [B_names[i] for i in B_idx],
    "B Condition":   [B_conditions[i] for i in B_idx],
    "B Row Value":   np.round(B[B_idx],4),
    "C Category":    [C_names[i] for i in C_idx],
    "C Condition":   [C_conditions[i] for i in C_idx],
    "C Row Value":   np.round(C[C_idx],4),
    "D Category":    [D_names[i] for i in D_idx],
    "D Condition":   [D_conditions[i] for i in D_idx],
    "D Row Value":   np.round(D[D_idx],4),
    "E Category":    [E_names[i] for i in E_idx],
    "E Condition":   [E_conditions[i] for i in E_idx],
    "E Row Value":   np.round(E[E_idx],4),
    "W1A":round(w1[0],4),"W1B":round(w1[1],4),"W1C":round(w1[2],4),
    "W1D":round(w1[3],4),"W1E":round(w1[4],4),
    "W2A":round(w2[0],4),"W2B":round(w2[1],4),"W2C":round(w2[2],4),
    "W2D":round(w2[3],4),"W2E":round(w2[4],4),
    "W3A":round(w3[0],4),"W3B":round(w3[1],4),"W3C":round(w3[2],4),
    "W3D":round(w3[3],4),"W3E":round(w3[4],4),
    "Score W1":    np.round(sc1,4),
    "Score W2":    np.round(sc2,4),
    "Score W3":    np.round(sc3,4),
    "Ro":          np.round(Ro,4),
    "Rpo":         np.round(Rpo,4),
    "Rso":         np.round(Rso,4),
    "Actual W1":   np.round(y1,4),
    "Actual W2":   np.round(y2,4),
    "Actual W3":   np.round(y3,4),
    "Actual Total":np.round(actual_total,4),
    "Predicted W1":    np.round(pw1,4),
    "Predicted W2":    np.round(pw2,4),
    "Predicted W3":    np.round(pw3,4),
    "Predicted Total": np.round(pt,4),
    "Split":       split_col,
})
out.to_csv(OUT_CSV,index=False)
print(f"\n  CSV saved → {len(out)} rows x {len(out.columns)} cols")

# =============================================================
# 13. BUILD EXCEL — same 5-sheet format
# =============================================================
print("STEP 6: Building Excel...")

cnt_A=pd.Series([A_names[i] for i in A_idx]).value_counts()
cnt_B=pd.Series([B_names[i] for i in B_idx]).value_counts()
cnt_C=pd.Series([C_names[i] for i in C_idx]).value_counts()
cnt_D=pd.Series([D_names[i] for i in D_idx]).value_counts()
cnt_E=pd.Series([E_names[i] for i in E_idx]).value_counts()
all_groups=[("A",A_names,A_conditions,A,cnt_A),
            ("B",B_names,B_conditions,B,cnt_B),
            ("C",C_names,C_conditions,C,cnt_C),
            ("D",D_names,D_conditions,D,cnt_D),
            ("E",E_names,E_conditions,E,cnt_E)]

HDR=PatternFill("solid",start_color="1F4E79")
GF={"A":PatternFill("solid",start_color="4472C4"),
    "B":PatternFill("solid",start_color="375623"),
    "C":PatternFill("solid",start_color="843C0C"),
    "D":PatternFill("solid",start_color="7B3F00"),
    "E":PatternFill("solid",start_color="4B0082")}
AF={"A":PatternFill("solid",start_color="D6E4F7"),
    "B":PatternFill("solid",start_color="E2EFDA"),
    "C":PatternFill("solid",start_color="FCE4D6"),
    "D":PatternFill("solid",start_color="F4CCAC"),
    "E":PatternFill("solid",start_color="E8D5F5")}
WH =PatternFill("solid",start_color="FFFFFF")
GRN=PatternFill("solid",start_color="E2EFDA")
RED=PatternFill("solid",start_color="FFDADA")
thin=Side(style="thin",color="BFBFBF")
BS  =Border(left=thin,right=thin,top=thin,bottom=thin)

def hc(ws,r,c,v,w=None):
    cl=ws.cell(r,c,v)
    cl.font=Font(bold=True,color="FFFFFF",size=11,name="Arial")
    cl.fill=HDR;cl.border=BS
    cl.alignment=Alignment(horizontal="center",vertical="center",wrap_text=True)
    if w: ws.column_dimensions[get_column_letter(c)].width=w

def dc(ws,r,c,v,fill=None,bold=False,align="center",fmt=None,fc="000000"):
    cl=ws.cell(r,c,v)
    cl.font=Font(bold=bold,size=10,name="Arial",color=fc)
    cl.fill=fill or WH;cl.border=BS
    cl.alignment=Alignment(horizontal=align,vertical="center",wrap_text=True)
    if fmt: cl.number_format=fmt

wb=Workbook()

# Sheet 1: Customer Output
ws1=wb.active; ws1.title="Customer Output"
ws1.freeze_panes="B2"; ws1.row_dimensions[1].height=30
for ci,col in enumerate(out.columns,1):
    hc(ws1,1,ci,col,w=max(12,min(35,len(col)+2)))
for ri,row in enumerate(out.itertuples(index=False),2):
    fill=AF["A"] if ri%2==0 else WH
    for ci,val in enumerate(row,1):
        dc(ws1,ri,ci,val,fill=fill)
    sc=list(out.columns).index("Split")+1
    ws1.cell(ri,sc).fill=GRN if row[sc-1]=="Train" else RED
print("  Sheet 1 (Customer Output) done")

# Sheet 2: Condition Summary
ws2=wb.create_sheet("Condition Summary")
ws2.freeze_panes="A2"; ws2.row_dimensions[1].height=35
for ci,(h,w) in enumerate(zip(
    ["Group","State","Optimized Value","Condition",
     "No. of Customers","% of Total"],
    [8,8,16,90,20,14]),1): hc(ws2,1,ci,h,w)
rn=2
for grp,names,conds,vals,cnt_ser in all_groups:
    for k,(nm,cond,val) in enumerate(zip(names,conds,vals)):
        cnt=int(cnt_ser.get(nm,0)); f=AF[grp] if k%2==0 else WH
        dc(ws2,rn,1,grp,fill=GF[grp],bold=True,fc="FFFFFF")
        dc(ws2,rn,2,nm, fill=f,bold=True)
        dc(ws2,rn,3,round(float(val),4),fill=f,fmt="0.0000")
        dc(ws2,rn,4,cond,fill=f,align="left")
        dc(ws2,rn,5,cnt, fill=f,fmt="#,##0")
        dc(ws2,rn,6,round(cnt/n,5),fill=f,fmt="0.00%")
        if cnt==0:
            ws2.cell(rn,5).font=Font(bold=True,color="FF0000",size=10)
        rn+=1
for ci,val in enumerate(["","TOTAL","","",n,1.0],1):
    c=ws2.cell(rn,ci,val)
    c.font=Font(bold=True,color="FFFFFF",size=11,name="Arial")
    c.fill=HDR;c.border=BS
    c.alignment=Alignment(horizontal="center",vertical="center")
ws2.cell(rn,5).number_format="#,##0"
ws2.cell(rn,6).number_format="0.00%"
print("  Sheet 2 (Condition Summary) done")

# Sheet 3: Bucket Values
ws3=wb.create_sheet("Bucket Values")
ws3.freeze_panes="A2"; ws3.row_dimensions[1].height=35
for ci,(h,w) in enumerate(zip(
    ["Group","Bucket","Optimized Value","Condition","Customers","% Total"],
    [8,8,16,90,15,12]),1): hc(ws3,1,ci,h,w)
rn=2
for grp,names,conds,vals,cnt_ser in all_groups:
    for k,(nm,cond,val) in enumerate(zip(names,conds,vals)):
        cnt=int(cnt_ser.get(nm,0)); f=AF[grp] if k%2==0 else WH
        dc(ws3,rn,1,grp,fill=GF[grp],bold=True,fc="FFFFFF")
        dc(ws3,rn,2,nm, fill=f,bold=True)
        dc(ws3,rn,3,round(float(val),4),fill=f,fmt="0.0000")
        dc(ws3,rn,4,cond,fill=f,align="left")
        dc(ws3,rn,5,cnt, fill=f,fmt="#,##0")
        dc(ws3,rn,6,round(cnt/n,5),fill=f,fmt="0.00%")
        if cnt==0:
            ws3.cell(rn,5).font=Font(bold=True,color="FF0000",size=10)
        rn+=1
print("  Sheet 3 (Bucket Values) done")

# Sheet 4: Weights
ws4=wb.create_sheet("Weights"); ws4.row_dimensions[1].height=30
for ci,(h,w) in enumerate(zip(
    ["Dimension","W1 (Interest Rate)","W2 (Admin Fee)",
     "W3 (Sourcing Fee)","W1 > 0.05?","W2 > 0.05?","W3 > 0.05?"],
    [28,18,18,18,14,14,14]),1): hc(ws4,1,ci,h,w)
dims=[("wA – Usage Type",         w1[0],w2[0],w3[0]),
      ("wB – Customer / Repayment",w1[1],w2[1],w3[1]),
      ("wC – Dealership / Source", w1[2],w2[2],w3[2]),
      ("wD – Branch Performance",  w1[3],w2[3],w3[3]),
      ("wE – Geography",           w1[4],w2[4],w3[4]),
      ("SUM",                      sum(w1),sum(w2),sum(w3))]
for ri,(lbl,v1,v2,v3) in enumerate(dims,2):
    is_s=(lbl=="SUM")
    f=HDR if is_s else (AF["B"] if ri%2==0 else WH)
    fc="FFFFFF" if is_s else "000000"
    for ci,val in enumerate(
        [lbl,round(float(v1),4),round(float(v2),4),round(float(v3),4)],1):
        c=ws4.cell(ri,ci,val)
        c.font=Font(bold=is_s,color=fc,size=10,name="Arial")
        c.fill=f;c.border=BS
        c.alignment=Alignment(
            horizontal="center" if ci>1 else "left",vertical="center")
        if ci>1: c.number_format="0.0000"
    if not is_s:
        for ci,v in [(5,v1),(6,v2),(7,v3)]:
            ok=v>=W_MIN-1e-9
            chk=f"✓ ({round(float(v),4)})" if ok else f"✗ ({round(float(v),4)})"
            c=ws4.cell(ri,ci,chk)
            c.font=Font(bold=True,
                       color="085041" if ok else "8B0000",
                       size=10,name="Arial")
            c.fill=GRN if ok else RED;c.border=BS
            c.alignment=Alignment(horizontal="center",vertical="center")
print("  Sheet 4 (Weights) done")

# Sheet 5: Validation
ws5=wb.create_sheet("Validation"); ws5.row_dimensions[1].height=30
for ci,(h,w) in enumerate(zip(["Metric","Train","Test"],[15,18,18]),1):
    hc(ws5,1,ci,h,w)
for ri,(m,tv,te) in enumerate(zip(["MSE","RMSE","MAE"],tr_m,te_m),2):
    f=AF["A"] if ri%2==0 else WH
    dc(ws5,ri,1,m,fill=f,bold=True,align="left")
    dc(ws5,ri,2,tv,fill=f,fmt="0.000000")
    dc(ws5,ri,3,te,fill=f,fmt="0.000000")
print("  Sheet 5 (Validation) done")

wb.save(OUT_EXCEL)
print(f"\n  Excel saved → {OUT_EXCEL}")

# =============================================================
# 14. FINAL SUMMARY
# =============================================================
print(f"\n{'='*65}")
print("FINAL SUMMARY — TRUST-CONSTR")
print(f"{'='*65}")
all_ok=True
for grp,names,_,_,cnt_ser in all_groups:
    used=sum(1 for nm in names if cnt_ser.get(nm,0)>0)
    total=len(names)
    missing=[nm for nm in names if cnt_ser.get(nm,0)==0]
    status="✓" if used==total else f"✗ MISSING: {missing}"
    print(f"  {grp}: {used}/{total}  {status}")
    if used!=total: all_ok=False

w_ok=all(w>=W_MIN-1e-9 for w in list(w1)+list(w2)+list(w3))
print(f"\n  States          : {'ALL 61 ✓' if all_ok else 'SOME MISSING'}")
print(f"  Weights > {W_MIN}   : {'ALL 15 ✓' if w_ok else '✗ SOME BELOW'}")
print(f"  W1 sum={sum(w1):.6f}  W2 sum={sum(w2):.6f}  W3 sum={sum(w3):.6f}")
print(f"  Train RMSE      : {tr_m[1]}")
print(f"  Test  RMSE      : {te_m[1]}")
print(f"\n  COMPARE WITH SLSQP:")
print(f"  SLSQP Test RMSE       = 0.01355")
print(f"  Trust-Constr Test RMSE = {te_m[1]}")
diff = round(0.01355 - te_m[1], 5)
if diff > 0:
    print(f"  Trust-Constr improved  = {diff:.5f} ✓ Better than SLSQP")
elif diff < 0:
    print(f"  SLSQP was better       = {abs(diff):.5f}")
else:
    print(f"  Same result — SLSQP already found the optimal solution")
print(f"\n  CSV   → {OUT_CSV}")
print(f"  Excel → {OUT_EXCEL}")
print(f"  Time  → {round(time.time()-t0,2)}s")
print(f"{'='*65}")
```

```python
"""
=============================================================
WAIVER OPTIMIZER — PARTICLE SWARM OPTIMIZATION (PSO)
=============================================================
"""

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from openpyxl import Workbook
from openpyxl.styles import Font, PatternFill, Alignment, Border, Side
from openpyxl.utils import get_column_letter
import time
import warnings
warnings.filterwarnings('ignore')

# ─────────────────────────────────────────────────────────
# PATHS
# ─────────────────────────────────────────────────────────
INPUT_CSV = r"C:\Users\intern-Sudhashri\Desktop\WAVIER\waiver_enriched_20000.csv"
OUT_EXCEL = r"C:\Users\intern-Sudhashri\Desktop\WAVIER\waiver_results_PSO.xlsx"
# ─────────────────────────────────────────────────────────

W_MIN = 0.05    # all weights must be >= this
t0    = time.time()

# =============================================================
# PSO PARAMETERS
# =============================================================
N_PARTICLES = 100    # number of particles in the swarm
N_ITER      = 200    # number of iterations
W_INERTIA   = 0.7   # inertia weight — how much to keep old velocity
C1          = 1.5   # cognitive weight — pull toward personal best
C2          = 1.5   # social weight   — pull toward global best
# Higher W_INERTIA = more exploration
# Higher C1        = particles trust their own best more
# Higher C2        = particles trust swarm best more

# =============================================================
# 1. LOAD AND ENRICH DATA
# =============================================================
print("STEP 1: Loading data...")
df = pd.read_csv(INPUT_CSV).copy()
n  = len(df)
np.random.seed(42)

# Enrich dealership columns
dlr_mask = df["Lead Source"].astype(str).str.strip().str.lower() == "dealership"
dlr_idx  = np.where(dlr_mask)[0]
n_high   = int(len(dlr_idx)*0.60)
hi_idx   = np.random.choice(dlr_idx,n_high,replace=False)
lo_idx   = np.setdiff1d(dlr_idx,hi_idx)
dq_new   = df["Dealership Quality (DQ)"].values.astype(float).copy()
dq_new[hi_idx]=np.random.uniform(80,100,n_high)
dq_new[lo_idx]=np.random.uniform(0,79,len(lo_idx))
df["Dealership Quality (DQ)"]=dq_new
lf_new=df["Lead Flow (LF%)"].values.astype(float).copy()
dp_new=df["Dealer Payout% (DP%)"].values.astype(float).copy()
lf_new[hi_idx]=np.random.uniform(0,100,n_high)
dp_new[hi_idx]=np.random.uniform(0,3,n_high)
df["Lead Flow (LF%)"]=lf_new
df["Dealer Payout% (DP%)"]=dp_new

# Enrich B columns
_rel_v  = df["Relationship with Company"].astype(str).values
_dpd_v  = df["Average Days Past Due (DPD) - Past 6 Months"].astype(float).values
_wvr_v  = df["Average Waiver Received Previously - Same Product"].astype(float).values
_prod_v = df["Previous Loan Product"].astype(str).str.strip().str.lower().values
prior_idx=np.where((_rel_v=="Existing Customer")&(_dpd_v<=5)&(_wvr_v>0))[0]
wvr_new=_wvr_v.copy(); p4=len(prior_idx)//4
wvr_new[prior_idx[:p4]]=np.random.uniform(0.01,0.49,p4)
wvr_new[prior_idx[p4:2*p4]]=np.random.uniform(0.5,0.99,p4)
wvr_new[prior_idx[2*p4:3*p4]]=np.random.uniform(1.0,1.49,p4)
wvr_new[prior_idx[3*p4:]]=np.random.uniform(1.5,2.0,len(prior_idx)-3*p4)
df["Average Waiver Received Previously - Same Product"]=wvr_new
DOH="Difference between Offered & Historical Interest Rate (Δoh)"
nps_idx=np.where((_rel_v=="Existing Customer")&(_dpd_v<=5)&
                  (wvr_new==0)&(_prod_v=="same product"))[0]
doh_new=df[DOH].astype(float).values.copy(); p5=len(nps_idx)//5
doh_new[nps_idx[:p5]]=np.random.uniform(-2,0,p5)
doh_new[nps_idx[p5:2*p5]]=np.random.uniform(0.01,0.49,p5)
doh_new[nps_idx[2*p5:3*p5]]=np.random.uniform(0.5,0.99,p5)
doh_new[nps_idx[3*p5:4*p5]]=np.random.uniform(1.0,1.49,p5)
doh_new[nps_idx[4*p5:]]=np.random.uniform(1.5,3.0,len(nps_idx)-4*p5)
df[DOH]=doh_new

print(f"  Loaded and enriched {n} rows in {round(time.time()-t0,2)}s")

# =============================================================
# 2. TARGETS AND RATES
# =============================================================
y1  = df["Actual Waiver Granted (Interest Rate Waiver)"].fillna(0).values.astype(np.float32)
y2  = df["Actual Waiver Granted (Admin & Processing Fee Waiver)"].fillna(0).values.astype(np.float32)
y3  = df["Actual Waiver Granted (Sourcing Fee Waiver)"].fillna(0).values.astype(np.float32)
Ro  = df["Interest Rate Offered (Ro)"].fillna(0).values.astype(np.float32)
Rpo = df["Admin & Processing Fee Offered (Rpo)"].fillna(0).values.astype(np.float32)
Rso = df["Sourcing Fee Offered (Rso)"].fillna(0).values.astype(np.float32)
idx_all   = np.arange(n)
train_idx, test_idx = train_test_split(idx_all, test_size=0.20, random_state=42)
actual_total = y1+y2+y3
print(f"  Train: {len(train_idx)}, Test: {len(test_idx)}")

# =============================================================
# 3. CONDITIONS
# =============================================================
A_names      = ["A1","A2"]
A_conditions = ["Usage Type = Personal","Usage Type = Commercial"]

B_names      = [f"B{i}" for i in range(1,13)]
B_conditions = [
    "New Customer",
    "Existing Customer, Bad Repayment Record (avg DPD last 6M > 5)",
    "Existing Customer, Good Repayment, Prior Waiver ∈ (0, 0.5)",
    "Existing Customer, Good Repayment, Prior Waiver ∈ [0.5, 1.0)",
    "Existing Customer, Good Repayment, Prior Waiver ∈ [1.0, 1.5)",
    "Existing Customer, Good Repayment, Prior Waiver ≥ 1.5",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product | Δoh ≤ 0",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product | Δoh ∈ (0, 0.5)",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product | Δoh ∈ [0.5, 1.0)",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product | Δoh ∈ [1.0, 1.5)",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product | Δoh ≥ 1.5",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Different Product",
]

C_names      = [f"C{i}" for i in range(1,39)]
C_conditions = ["Source ≠ Dealership","Source = Dealership, DQ < 80%"]
for lf_lbl in ["LF < 10%","10% ≤ LF < 20%","20% ≤ LF < 30%",
                "30% ≤ LF < 40%","40% ≤ LF < 50%","LF ≥ 50%"]:
    for dp_lbl in ["DP < 0.3%","0.3% ≤ DP < 0.6%","0.6% ≤ DP < 0.9%",
                   "0.9% ≤ DP < 1.2%","1.2% ≤ DP < 1.5%","DP ≥ 1.5%"]:
        C_conditions.append(f"Source = Dealership, DQ ≥ 80%, {lf_lbl}, {dp_lbl}")

D_names      = ["D1","D2","D3","D4"]
D_conditions = [
    "Branch Ahead of Profitability Target & Behind Disbursal Target",
    "Branch Ahead of Profitability Target & Ahead of Disbursal Target",
    "Branch Behind Profitability Target & Behind Disbursal Target",
    "Branch Behind Profitability Target & Ahead of Disbursal Target",
]

E_names      = [f"E{i}" for i in range(1,6)]
E_conditions = [
    "Category 1 – Tier 1 Urban","Category 2 – Tier 1 Rural",
    "Category 3 – Tier 2 Urban","Category 4 – Tier 2 Rural",
    "Category 5 – Tier 3 Rural & Others",
]

# =============================================================
# 4. BUCKET ASSIGNMENT (rule-based — same as SLSQP)
# =============================================================
print("STEP 2: Assigning buckets A-E...")

A_idx = np.where(df["Usage Type"].eq("Personal"),0,1).astype(np.int32)

_rel  = df["Relationship with Company"].astype(str).values
_dpd  = df["Average Days Past Due (DPD) - Past 6 Months"].astype(float).values
_wvr  = df["Average Waiver Received Previously - Same Product"].astype(float).values
_doh  = df[DOH].astype(float).values
_prod = df["Previous Loan Product"].astype(str).str.strip().str.lower().values
is_new=(_rel=="New Customer"); is_bad=(~is_new)&(_dpd>5); is_good=(~is_new)&(_dpd<=5)
prior=is_good&(_wvr>0); nps=is_good&(_wvr==0)&(_prod=="same product")
diffp=is_good&(_wvr==0)&(_prod=="different")
B_idx=np.zeros(n,dtype=np.int32)
B_idx[is_new]=0;B_idx[is_bad]=1
B_idx[prior&(_wvr<0.5)]=2;B_idx[prior&(_wvr>=0.5)&(_wvr<1.0)]=3
B_idx[prior&(_wvr>=1.0)&(_wvr<1.5)]=4;B_idx[prior&(_wvr>=1.5)]=5
B_idx[nps&(_doh<=0)]=6;B_idx[nps&(_doh>0)&(_doh<0.5)]=7
B_idx[nps&(_doh>=0.5)&(_doh<1.0)]=8;B_idx[nps&(_doh>=1.0)&(_doh<1.5)]=9
B_idx[nps&(_doh>=1.5)]=10;B_idx[diffp]=11

_src=df["Lead Source"].astype(str).str.strip().str.lower().values
_dq=df["Dealership Quality (DQ)"].astype(float).values
_lf=df["Lead Flow (LF%)"].astype(float).values
_dp=df["Dealer Payout% (DP%)"].astype(float).values
is_dlr=(_src=="dealership"); dlr_ok=is_dlr&(_dq>=80)
lf_grp=np.where(_lf<10,0,np.where(_lf<20,1,np.where(_lf<30,2,
        np.where(_lf<40,3,np.where(_lf<50,4,5))))).astype(np.int32)
dp_grp=np.where(_dp<0.3,0,np.where(_dp<0.6,1,np.where(_dp<0.9,2,
        np.where(_dp<1.2,3,np.where(_dp<1.5,4,5))))).astype(np.int32)
C_idx=np.zeros(n,dtype=np.int32)
C_idx[~is_dlr]=0;C_idx[is_dlr&(_dq<80)]=1
C_idx[dlr_ok]=2+lf_grp[dlr_ok]*6+dp_grp[dlr_ok]

_profit=df["Tracking towards profitability target"].astype(str).values
_disb  =df["Tracking towards disbursal target"].astype(str).values
D_idx  =np.full(n,2,dtype=np.int32)
D_idx[(_profit=="Ahead")&(_disb=="Behind")]=0
D_idx[(_profit=="Ahead")&(_disb=="Ahead")] =1
D_idx[(_profit=="Behind")&(_disb=="Ahead")]=3

_ecat=df["Branch Competitive Category"].astype(str).str.strip().values
E_idx=np.zeros(n,dtype=np.int32)
for v,ei in [("1",0),("2",1),("3",2),("4",3),("5",4)]:
    E_idx[pd.Series(_ecat).str.contains(v,regex=False).values]=ei

print(f"  A:{len(set(A_idx))}/2  B:{len(set(B_idx))}/12  "
      f"C:{len(set(C_idx))}/38  D:{len(set(D_idx))}/4  E:{len(set(E_idx))}/5")

# =============================================================
# 5. ONE-HOT MATRICES
# =============================================================
def make_oh(ia,nb,rows):
    M=np.zeros((len(rows),nb),dtype=np.float32)
    M[np.arange(len(rows)),ia[rows]]=1.0
    return M

tr=train_idx
OA=make_oh(A_idx,2,tr);   OA_all=make_oh(A_idx,2,idx_all)
OB=make_oh(B_idx,12,tr);  OB_all=make_oh(B_idx,12,idx_all)
OC=make_oh(C_idx,38,tr);  OC_all=make_oh(C_idx,38,idx_all)
OD=make_oh(D_idx,4,tr);   OD_all=make_oh(D_idx,4,idx_all)
OE=make_oh(E_idx,5,tr);   OE_all=make_oh(E_idx,5,idx_all)
y1_tr=y1[tr];y2_tr=y2[tr];y3_tr=y3[tr]
s1=0.1*Ro[tr];s2=0.1*Rpo[tr];s3=0.1*Rso[tr]

print(f"  One-hot matrices built in {round(time.time()-t0,2)}s")

# =============================================================
# 6. OBJECTIVE WITH PENALTY (for PSO — no hard constraints)
# =============================================================
PENALTY = 1000   # penalty strength for constraint violations

def objective_pso(p):
    """
    Same loss as SLSQP but with penalty terms for business rules.
    PSO cannot enforce hard constraints so we add penalty instead.
    """
    A_=p[0:2].astype(np.float32);  B_=p[2:14].astype(np.float32)
    C_=p[14:52].astype(np.float32); D_=p[52:56].astype(np.float32)
    E_=p[56:61].astype(np.float32)
    w1_=p[61:66]; w2_=p[66:71]; w3_=p[71:76]

    # Prediction error
    Av=OA@A_;Bv=OB@B_;Cv=OC@C_;Dv=OD@D_;Ev=OE@E_
    sc1=w1_[0]*Av+w1_[1]*Bv+w1_[2]*Cv+w1_[3]*Dv+w1_[4]*Ev
    sc2=w2_[0]*Av+w2_[1]*Bv+w2_[2]*Cv+w2_[3]*Dv+w2_[4]*Ev
    sc3=w3_[0]*Av+w3_[1]*Bv+w3_[2]*Cv+w3_[3]*Dv+w3_[4]*Ev
    r1=y1_tr-sc1*s1; r2=y2_tr-sc2*s2; r3=y3_tr-sc3*s3
    base_loss = float(r1@r1+r2@r2+r3@r3+0.001*np.dot(p,p))

    # Penalty terms for business rules
    penalty = 0.0

    # A1 > A2 by at least 0.5
    if p[0]-p[1] < 0.5:
        penalty += PENALTY*(0.5-(p[0]-p[1]))**2

    # B1 = 5
    penalty += PENALTY*(p[2]-5)**2

    # B2 = 0
    penalty += PENALTY*(p[3])**2

    # B3 < B4 < B5 < B6 (ordering)
    if p[5]-p[4] < 0.1: penalty += PENALTY*(0.1-(p[5]-p[4]))**2
    if p[6]-p[5] < 0.1: penalty += PENALTY*(0.1-(p[6]-p[5]))**2
    if p[7]-p[6] < 0.1: penalty += PENALTY*(0.1-(p[7]-p[6]))**2

    # B12 = 5
    penalty += PENALTY*(p[13]-5)**2

    # C1 = 5
    penalty += PENALTY*(p[14]-5)**2

    # C2 = 0
    penalty += PENALTY*(p[15])**2

    # D1 > D2 > D3 > D4 (ordering)
    if p[52]-p[53] < 0.1: penalty += PENALTY*(0.1-(p[52]-p[53]))**2
    if p[53]-p[54] < 0.1: penalty += PENALTY*(0.1-(p[53]-p[54]))**2
    if p[54]-p[55] < 0.1: penalty += PENALTY*(0.1-(p[54]-p[55]))**2

    # D4 = 0
    penalty += PENALTY*(p[55])**2

    # E1 > E2 > E3 > E4 > E5 (ordering)
    if p[56]-p[57] < 0.1: penalty += PENALTY*(0.1-(p[56]-p[57]))**2
    if p[57]-p[58] < 0.1: penalty += PENALTY*(0.1-(p[57]-p[58]))**2
    if p[58]-p[59] < 0.1: penalty += PENALTY*(0.1-(p[58]-p[59]))**2
    if p[59]-p[60] < 0.1: penalty += PENALTY*(0.1-(p[59]-p[60]))**2

    # E5 = 1
    penalty += PENALTY*(p[60]-1)**2

    # Weights sum to 1
    penalty += PENALTY*(np.sum(w1_)-1)**2
    penalty += PENALTY*(np.sum(w2_)-1)**2
    penalty += PENALTY*(np.sum(w3_)-1)**2

    return base_loss + penalty

# =============================================================
# 7. BOUNDS FOR PSO
# =============================================================
# Lower and upper bounds for each of the 76 parameters
lb = np.array([0.0]*61 + [W_MIN]*15)   # lower bounds
ub = np.array([10.0]*61 + [1.0]*15)    # upper bounds

# =============================================================
# 8. PSO IMPLEMENTATION (from scratch — no extra library needed)
# =============================================================
print(f"\nSTEP 3: Running PSO...")
print(f"  Particles : {N_PARTICLES}")
print(f"  Iterations: {N_ITER}")
print(f"  Dimensions: 76 parameters")
print()

DIM = 76   # 76 parameters to optimize

# Initialize particles
# Each particle has a position (76 values) and velocity (76 values)
np.random.seed(0)
positions  = lb + np.random.rand(N_PARTICLES, DIM) * (ub - lb)
velocities = np.zeros((N_PARTICLES, DIM))

# Fix initial positions to respect weight constraints approximately
for i in range(N_PARTICLES):
    # Normalize W1, W2, W3 weights to sum to 1
    positions[i,61:66] = np.abs(positions[i,61:66])
    positions[i,61:66] = np.maximum(positions[i,61:66], W_MIN)
    positions[i,61:66] /= positions[i,61:66].sum()

    positions[i,66:71] = np.abs(positions[i,66:71])
    positions[i,66:71] = np.maximum(positions[i,66:71], W_MIN)
    positions[i,66:71] /= positions[i,66:71].sum()

    positions[i,71:76] = np.abs(positions[i,71:76])
    positions[i,71:76] = np.maximum(positions[i,71:76], W_MIN)
    positions[i,71:76] /= positions[i,71:76].sum()

# Evaluate initial fitness for all particles
fitness = np.array([objective_pso(positions[i]) for i in range(N_PARTICLES)])

# Personal best = each particle's own best position and fitness
personal_best_pos = positions.copy()
personal_best_fit = fitness.copy()

# Global best = best position found by any particle
global_best_idx = np.argmin(fitness)
global_best_pos = positions[global_best_idx].copy()
global_best_fit = fitness[global_best_idx]

print(f"  Initial best loss = {global_best_fit:.4f}")

# PSO main loop
pso_log = []
for it in range(N_ITER):
    # Random factors for velocity update
    r1 = np.random.rand(N_PARTICLES, DIM)   # random toward personal best
    r2 = np.random.rand(N_PARTICLES, DIM)   # random toward global best

    # Update velocity for each particle
    # velocity = inertia×old_velocity
    #          + cognitive×(personal_best - current)
    #          + social×(global_best - current)
    velocities = (W_INERTIA * velocities
                  + C1 * r1 * (personal_best_pos - positions)
                  + C2 * r2 * (global_best_pos   - positions))

    # Update position
    positions = positions + velocities

    # Clip positions to stay within bounds
    positions = np.clip(positions, lb, ub)

    # Fix weight sums to 1 after clipping
    for i in range(N_PARTICLES):
        for w_slice in [slice(61,66), slice(66,71), slice(71,76)]:
            positions[i,w_slice] = np.maximum(positions[i,w_slice], W_MIN)
            s = positions[i,w_slice].sum()
            if s > 0:
                positions[i,w_slice] /= s

    # Evaluate fitness for all particles
    fitness = np.array([objective_pso(positions[i]) for i in range(N_PARTICLES)])

    # Update personal best
    improved = fitness < personal_best_fit
    personal_best_pos[improved] = positions[improved].copy()
    personal_best_fit[improved] = fitness[improved]

    # Update global best
    best_idx = np.argmin(personal_best_fit)
    if personal_best_fit[best_idx] < global_best_fit:
        global_best_fit = personal_best_fit[best_idx]
        global_best_pos = personal_best_pos[best_idx].copy()

    pso_log.append(round(global_best_fit,4))

    # Print progress every 20 iterations
    if (it+1) % 20 == 0:
        print(f"  Iteration {it+1:>4}/{N_ITER} | Best loss = {global_best_fit:.6f}")

print(f"\n  PSO done in {round(time.time()-t0,2)}s")
print(f"  Final best loss = {global_best_fit:.6f}")

# =============================================================
# 9. POST-PROCESSING — fix hard rules exactly
# =============================================================
print("\nSTEP 4: Post-processing — fixing business rules exactly...")

p = global_best_pos.copy()

# Fix B1=5, B2=0, B12=5, C1=5, C2=0, D4=0, E5=1
p[2]  = 5.0   # B1
p[3]  = 0.0   # B2
p[13] = 5.0   # B12
p[14] = 5.0   # C1
p[15] = 0.0   # C2
p[55] = 0.0   # D4
p[60] = 1.0   # E5

# Fix weight sums exactly to 1
for sl in [slice(61,66), slice(66,71), slice(71,76)]:
    p[sl] = np.maximum(p[sl], W_MIN)
    p[sl] = p[sl] / p[sl].sum()

# Fix A1 > A2
if p[0] <= p[1]:
    p[0] = p[1] + 0.5

# Fix ordering B3 < B4 < B5 < B6
for i in [5,6,7]:
    if p[i] <= p[i-1]:
        p[i] = p[i-1] + 0.1

# Fix ordering D1 > D2 > D3 > D4
for i in [53,54,55]:
    if p[i] >= p[i-1]:
        p[i] = p[i-1] - 0.1
        p[i] = max(p[i], 0)

# Fix ordering E1 > E2 > E3 > E4 > E5
for i in [57,58,59,60]:
    if p[i] >= p[i-1]:
        p[i] = p[i-1] - 0.1
        p[i] = max(p[i], 0)

A  = p[0:2];  B  = p[2:14];  C  = p[14:52]
D  = p[52:56]; E = p[56:61]
w1 = p[61:66]; w2 = p[66:71]; w3 = p[71:76]

print(f"  A1={A[0]:.4f}  A2={A[1]:.4f}")
print(f"  B1={B[0]:.4f}(fixed)  B2={B[1]:.4f}(fixed)  "
      f"B3={B[2]:.4f}  B4={B[3]:.4f}  B5={B[4]:.4f}  B6={B[5]:.4f}")
print(f"  D1={D[0]:.4f}  D2={D[1]:.4f}  D3={D[2]:.4f}  D4={D[3]:.4f}(fixed)")
print(f"  W1={np.round(w1,4)}  sum={sum(w1):.4f}  "
      f"min={min(w1):.4f}  {'✓' if min(w1)>=W_MIN-1e-9 else '✗'}")
print(f"  W2={np.round(w2,4)}  sum={sum(w2):.4f}  "
      f"min={min(w2):.4f}  {'✓' if min(w2)>=W_MIN-1e-9 else '✗'}")
print(f"  W3={np.round(w3,4)}  sum={sum(w3):.4f}  "
      f"min={min(w3):.4f}  {'✓' if min(w3)>=W_MIN-1e-9 else '✗'}")

# =============================================================
# 10. PREDICT
# =============================================================
Av=OA_all@A;Bv=OB_all@B;Cv=OC_all@C;Dv=OD_all@D;Ev=OE_all@E
sc1=w1[0]*Av+w1[1]*Bv+w1[2]*Cv+w1[3]*Dv+w1[4]*Ev
sc2=w2[0]*Av+w2[1]*Bv+w2[2]*Cv+w2[3]*Dv+w2[4]*Ev
sc3=w3[0]*Av+w3[1]*Bv+w3[2]*Cv+w3[3]*Dv+w3[4]*Ev
pw1=sc1*0.1*Ro; pw2=sc2*0.1*Rpo; pw3=sc3*0.1*Rso
pt=pw1+pw2+pw3

def met(a,q):
    mse=float(mean_squared_error(a,q))
    return round(mse,6),round(float(np.sqrt(mse)),6),\
           round(float(np.mean(np.abs(a-q))),6)

tr_m=met(actual_total[train_idx],pt[train_idx])
te_m=met(actual_total[test_idx], pt[test_idx])

print(f"\n  Train → MSE={tr_m[0]}  RMSE={tr_m[1]}  MAE={tr_m[2]}")
print(f"  Test  → MSE={te_m[0]}  RMSE={te_m[1]}  MAE={te_m[2]}")

# =============================================================
# 11. BUILD EXCEL — same format as SLSQP
# =============================================================
print("\nSTEP 5: Building Excel...")

cnt_A=pd.Series([A_names[i] for i in A_idx]).value_counts()
cnt_B=pd.Series([B_names[i] for i in B_idx]).value_counts()
cnt_C=pd.Series([C_names[i] for i in C_idx]).value_counts()
cnt_D=pd.Series([D_names[i] for i in D_idx]).value_counts()
cnt_E=pd.Series([E_names[i] for i in E_idx]).value_counts()
all_groups=[("A",A_names,A_conditions,A,cnt_A),
            ("B",B_names,B_conditions,B,cnt_B),
            ("C",C_names,C_conditions,C,cnt_C),
            ("D",D_names,D_conditions,D,cnt_D),
            ("E",E_names,E_conditions,E,cnt_E)]

HDR=PatternFill("solid",start_color="1F4E79")
GF={"A":PatternFill("solid",start_color="4472C4"),
    "B":PatternFill("solid",start_color="375623"),
    "C":PatternFill("solid",start_color="843C0C"),
    "D":PatternFill("solid",start_color="7B3F00"),
    "E":PatternFill("solid",start_color="4B0082")}
AF={"A":PatternFill("solid",start_color="D6E4F7"),
    "B":PatternFill("solid",start_color="E2EFDA"),
    "C":PatternFill("solid",start_color="FCE4D6"),
    "D":PatternFill("solid",start_color="F4CCAC"),
    "E":PatternFill("solid",start_color="E8D5F5")}
WH =PatternFill("solid",start_color="FFFFFF")
GRN=PatternFill("solid",start_color="E2EFDA")
RED=PatternFill("solid",start_color="FFDADA")
thin=Side(style="thin",color="BFBFBF")
BS  =Border(left=thin,right=thin,top=thin,bottom=thin)

def hc(ws,r,c,v,w=None):
    cl=ws.cell(r,c,v)
    cl.font=Font(bold=True,color="FFFFFF",size=11,name="Arial")
    cl.fill=HDR;cl.border=BS
    cl.alignment=Alignment(horizontal="center",vertical="center",wrap_text=True)
    if w: ws.column_dimensions[get_column_letter(c)].width=w

def dc(ws,r,c,v,fill=None,bold=False,align="center",fmt=None,fc="000000"):
    cl=ws.cell(r,c,v)
    cl.font=Font(bold=bold,size=10,name="Arial",color=fc)
    cl.fill=fill or WH;cl.border=BS
    cl.alignment=Alignment(horizontal=align,vertical="center",wrap_text=True)
    if fmt: cl.number_format=fmt

wb=Workbook()

# Sheet 1: Customer Output
ws1=wb.active; ws1.title="Customer Output"
ws1.freeze_panes="B2"; ws1.row_dimensions[1].height=30
cols=["Customer","A Category","A Condition","A Row Value",
      "B Category","B Condition","B Row Value",
      "C Category","C Condition","C Row Value",
      "D Category","D Condition","D Row Value",
      "E Category","E Condition","E Row Value",
      "W1A","W1B","W1C","W1D","W1E",
      "W2A","W2B","W2C","W2D","W2E",
      "W3A","W3B","W3C","W3D","W3E",
      "Score W1","Score W2","Score W3",
      "Ro","Rpo","Rso",
      "Actual W1","Actual W2","Actual W3","Actual Total",
      "Predicted W1","Predicted W2","Predicted W3","Predicted Total"]
for ci,col in enumerate(cols,1):
    hc(ws1,1,ci,col,w=max(10,min(30,len(col)+2)))
for ri in range(n):
    f=AF["A"] if ri%2==0 else WH
    row_vals=[
        df["Name of Customer"].iloc[ri],
        A_names[A_idx[ri]],A_conditions[A_idx[ri]],round(float(A[A_idx[ri]]),4),
        B_names[B_idx[ri]],B_conditions[B_idx[ri]],round(float(B[B_idx[ri]]),4),
        C_names[C_idx[ri]],C_conditions[C_idx[ri]],round(float(C[C_idx[ri]]),4),
        D_names[D_idx[ri]],D_conditions[D_idx[ri]],round(float(D[D_idx[ri]]),4),
        E_names[E_idx[ri]],E_conditions[E_idx[ri]],round(float(E[E_idx[ri]]),4),
        round(float(w1[0]),4),round(float(w1[1]),4),round(float(w1[2]),4),
        round(float(w1[3]),4),round(float(w1[4]),4),
        round(float(w2[0]),4),round(float(w2[1]),4),round(float(w2[2]),4),
        round(float(w2[3]),4),round(float(w2[4]),4),
        round(float(w3[0]),4),round(float(w3[1]),4),round(float(w3[2]),4),
        round(float(w3[3]),4),round(float(w3[4]),4),
        round(float(sc1[ri]),4),round(float(sc2[ri]),4),round(float(sc3[ri]),4),
        round(float(Ro[ri]),4),round(float(Rpo[ri]),4),round(float(Rso[ri]),4),
        round(float(y1[ri]),4),round(float(y2[ri]),4),round(float(y3[ri]),4),
        round(float(actual_total[ri]),4),
        round(float(pw1[ri]),4),round(float(pw2[ri]),4),round(float(pw3[ri]),4),
        round(float(pt[ri]),4),
    ]
    for ci,val in enumerate(row_vals,1):
        dc(ws1,ri+2,ci,val,fill=f)
print("  Sheet 1 (Customer Output) done")

# Sheet 2: Condition Summary
ws2=wb.create_sheet("Condition Summary")
ws2.freeze_panes="A2"; ws2.row_dimensions[1].height=35
for ci,(h,w) in enumerate(zip(
    ["Group","State","Optimized Value","Condition",
     "No. of Customers","% of Total"],
    [8,8,16,90,20,14]),1): hc(ws2,1,ci,h,w)
rn=2
for grp,names,conds,vals,cnt_ser in all_groups:
    for k,(nm,cond,val) in enumerate(zip(names,conds,vals)):
        cnt=int(cnt_ser.get(nm,0)); f=AF[grp] if k%2==0 else WH
        dc(ws2,rn,1,grp,fill=GF[grp],bold=True,fc="FFFFFF")
        dc(ws2,rn,2,nm, fill=f,bold=True)
        dc(ws2,rn,3,round(float(val),4),fill=f,fmt="0.0000")
        dc(ws2,rn,4,cond,fill=f,align="left")
        dc(ws2,rn,5,cnt, fill=f,fmt="#,##0")
        dc(ws2,rn,6,round(cnt/n,5),fill=f,fmt="0.00%")
        if cnt==0:
            ws2.cell(rn,5).font=Font(bold=True,color="FF0000",size=10)
        rn+=1
for ci,val in enumerate(["","TOTAL","","",n,1.0],1):
    c=ws2.cell(rn,ci,val)
    c.font=Font(bold=True,color="FFFFFF",size=11,name="Arial")
    c.fill=HDR;c.border=BS
    c.alignment=Alignment(horizontal="center",vertical="center")
ws2.cell(rn,5).number_format="#,##0"
ws2.cell(rn,6).number_format="0.00%"
print("  Sheet 2 (Condition Summary) done")

# Sheet 3: Bucket Values
ws3=wb.create_sheet("Bucket Values")
ws3.freeze_panes="A2"; ws3.row_dimensions[1].height=35
for ci,(h,w) in enumerate(zip(
    ["Group","Bucket","Optimized Value","Condition","Customers","% Total"],
    [8,8,16,90,15,12]),1): hc(ws3,1,ci,h,w)
rn=2
for grp,names,conds,vals,cnt_ser in all_groups:
    for k,(nm,cond,val) in enumerate(zip(names,conds,vals)):
        cnt=int(cnt_ser.get(nm,0)); f=AF[grp] if k%2==0 else WH
        dc(ws3,rn,1,grp,fill=GF[grp],bold=True,fc="FFFFFF")
        dc(ws3,rn,2,nm, fill=f,bold=True)
        dc(ws3,rn,3,round(float(val),4),fill=f,fmt="0.0000")
        dc(ws3,rn,4,cond,fill=f,align="left")
        dc(ws3,rn,5,cnt, fill=f,fmt="#,##0")
        dc(ws3,rn,6,round(cnt/n,5),fill=f,fmt="0.00%")
        if cnt==0:
            ws3.cell(rn,5).font=Font(bold=True,color="FF0000",size=10)
        rn+=1
print("  Sheet 3 (Bucket Values) done")

# Sheet 4: Weights
ws4=wb.create_sheet("Weights"); ws4.row_dimensions[1].height=30
for ci,(h,w) in enumerate(zip(
    ["Dimension","W1 (Interest Rate)","W2 (Admin Fee)",
     "W3 (Sourcing Fee)","W1 > 0.05?","W2 > 0.05?","W3 > 0.05?"],
    [28,18,18,18,14,14,14]),1): hc(ws4,1,ci,h,w)
dims=[("wA – Usage Type",        w1[0],w2[0],w3[0]),
      ("wB – Customer / Repayment",w1[1],w2[1],w3[1]),
      ("wC – Dealership / Source", w1[2],w2[2],w3[2]),
      ("wD – Branch Performance",  w1[3],w2[3],w3[3]),
      ("wE – Geography",           w1[4],w2[4],w3[4]),
      ("SUM",                      sum(w1),sum(w2),sum(w3))]
for ri,(lbl,v1,v2,v3) in enumerate(dims,2):
    is_s=(lbl=="SUM")
    f=HDR if is_s else (AF["B"] if ri%2==0 else WH)
    fc="FFFFFF" if is_s else "000000"
    for ci,val in enumerate(
        [lbl,round(float(v1),4),round(float(v2),4),round(float(v3),4)],1):
        c=ws4.cell(ri,ci,val)
        c.font=Font(bold=is_s,color=fc,size=10,name="Arial")
        c.fill=f;c.border=BS
        c.alignment=Alignment(
            horizontal="center" if ci>1 else "left",vertical="center")
        if ci>1: c.number_format="0.0000"
    if not is_s:
        for ci,v in [(5,v1),(6,v2),(7,v3)]:
            ok=v>=W_MIN-1e-9
            chk=f"✓ ({round(float(v),4)})" if ok else f"✗ ({round(float(v),4)})"
            c=ws4.cell(ri,ci,chk)
            c.font=Font(bold=True,
                       color="085041" if ok else "8B0000",
                       size=10,name="Arial")
            c.fill=GRN if ok else RED;c.border=BS
            c.alignment=Alignment(horizontal="center",vertical="center")
print("  Sheet 4 (Weights) done")

# Sheet 5: Validation
ws5=wb.create_sheet("Validation"); ws5.row_dimensions[1].height=30
for ci,(h,w) in enumerate(zip(["Metric","Train","Test"],[15,18,18]),1):
    hc(ws5,1,ci,h,w)
for ri,(m,tv,te) in enumerate(zip(["MSE","RMSE","MAE"],tr_m,te_m),2):
    f=AF["A"] if ri%2==0 else WH
    dc(ws5,ri,1,m,fill=f,bold=True,align="left")
    dc(ws5,ri,2,tv,fill=f,fmt="0.000000")
    dc(ws5,ri,3,te,fill=f,fmt="0.000000")
print("  Sheet 5 (Validation) done")

wb.save(OUT_EXCEL)
print(f"\n  Excel saved → {OUT_EXCEL}")

# =============================================================
# 12. FINAL SUMMARY
# =============================================================
print(f"\n{'='*65}")
print("PSO COMPLETE")
print(f"{'='*65}")
print(f"  Particles         : {N_PARTICLES}")
print(f"  Iterations        : {N_ITER}")
print(f"  Final loss        : {global_best_fit:.6f}")
print(f"  Train RMSE        : {tr_m[1]}")
print(f"  Test  RMSE        : {te_m[1]}")
wt_ok=all(w>=W_MIN-1e-9 for w in list(w1)+list(w2)+list(w3))
print(f"  Weights > {W_MIN}    : {'✓ All 15' if wt_ok else '✗ Some below'}")
print(f"  B1={B[0]:.2f}(fixed)  B2={B[1]:.2f}(fixed)  "
      f"D4={D[3]:.2f}(fixed)  E5={E[4]:.2f}(fixed)")
print(f"\n  COMPARE WITH SLSQP:")
print(f"  SLSQP Test RMSE  = 0.01355")
print(f"  PSO   Test RMSE  = {te_m[1]}")
diff=round(0.01355-te_m[1],5)
if diff>0:
    print(f"  PSO improved by  = {diff:.5f} ✓")
elif diff<0:
    print(f"  SLSQP was better = {abs(diff):.5f}")
else:
    print(f"  Same result — SLSQP already found global optimum")
print(f"\n  Excel → {OUT_EXCEL}")
print(f"  Time  → {round(time.time()-t0,2)}s")
print(f"{'='*65}")
```
