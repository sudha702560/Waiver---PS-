```python

# STEP 1 — LOAD DATA
print("STEP 1")
 
df = pd.read_csv(INPUT_CSV).copy()
n  = len(df)
np.random.seed(42)
 
print(f"  Loaded {n} rows x {df.shape[1]} columns")
 
# columns we need
y1  = df["Actual Waiver Granted (Interest Rate Waiver)"].fillna(0).values.astype(np.float32)
y2  = df["Actual Waiver Granted (Admin & Processing Fee Waiver)"].fillna(0).values.astype(np.float32)
y3  = df["Actual Waiver Granted (Sourcing Fee Waiver)"].fillna(0).values.astype(np.float32)

Ro  = df["Interest Rate Offered (Ro)"].fillna(0).values.astype(np.float32)
Rpo = df["Admin & Processing Fee Offered (Rpo)"].fillna(0).values.astype(np.float32)
Rso = df["Sourcing Fee Offered (Rso)"].fillna(0).values.astype(np.float32)
 
idx_all   = np.arange(n)
train_idx, test_idx = train_test_split(idx_all, test_size=0.20, random_state=42)
 
print(f"  Train: {len(train_idx)} customers")
print(f"  Test : {len(test_idx)} customers")
```

```python
print(" 3: Group B")
 
B_names      = [f"B{i}" for i in range(1, 13)]
B_conditions = [
    "New Customer",
    "Existing Customer, Bad Repayment (DPD > 5)",
    "Existing Customer ,DPD ≤ 5, Good Repayment, Prior Waiver ∈ (0, 0.5)",
    "Existing Customer,DPD ≤ 5, Good Repayment, Prior Waiver ∈ [0.5, 1.0)",
    "Existing Customer,DPD ≤ 5, Good Repayment, Prior Waiver ∈ [1.0, 1.5)",
    "Existing Customer,DPD ≤ 5, Good Repayment, Prior Waiver ≥ 1.5",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product, Δoh ≤ 0",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product, Δoh ∈ (0, 0.5)",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product, Δoh ∈ [0.5, 1.0)",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product, Δoh ∈ [1.0, 1.5)",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Same Product, Δoh ≥ 1.5",
    "Existing Customer, DPD ≤ 5, No Prior Waiver, Different Product",
]

 
# columns
_rel  = df["Relationship with Company"].astype(str).values
_dpd  = df["Average Days Past Due (DPD) - Past 6 Months"].astype(float).values
_wvr  = df["Average Waiver Received Previously - Same Product"].astype(float).values
_doh  = df["Difference between Offered & Historical Interest Rate (Δoh)"].astype(float).values
_prod = df["Previous Loan Product"].astype(str).str.strip().str.lower().values
 
#  Customer 
is_new  = (_rel == "New Customer")
is_bad  = (~is_new) & (_dpd > 5)
is_good = (~is_new) & (_dpd <= 5)
 
# good customers
prior = is_good & (_wvr > 0)#Good customer and  already received waiver before
nps   = is_good & (_wvr == 0) & (_prod == "same product")#Good customer,no previous waiver,Same product
diffp = is_good & (_wvr == 0) & (_prod == "different")
 #Good customer,No previous ,waiverDifferent product
# Assign bucket index
B_idx = np.zeros(n, dtype=np.int32)

B_idx[is_new] = 0   # B1
B_idx[is_bad] = 1   # B2
B_idx[diffp]  = 11  # B12
 
# Prior waiver bands — explicit conditions
B_idx[prior & (_wvr < 0.5)]                  = 2   # B3
B_idx[prior & (_wvr >= 0.5) & (_wvr < 1.0)]  = 3   # B4
B_idx[prior & (_wvr >= 1.0) & (_wvr < 1.5)]  = 4   # B5
B_idx[prior & (_wvr >= 1.5)]                 = 5   # B6
 
# Delta oh bands — explicit conditions
B_idx[nps & (_doh <= 0)]                      = 6   # B7
B_idx[nps & (_doh > 0)   & (_doh < 0.5)]     = 7   # B8
B_idx[nps & (_doh >= 0.5) & (_doh < 1.0)]    = 8   # B9
B_idx[nps & (_doh >= 1.0) & (_doh < 1.5)]    = 9   # B10
B_idx[nps & (_doh >= 1.5)]                   = 10  # B11
 
# Print result
print(f"  {'Bucket':<6} {'Customers':>10}  Condition")
print(f"  {'─'*60}")
for i, (nm, cond) in enumerate(zip(B_names, B_conditions)):
    cnt = (B_idx == i).sum()
    print(f"  {nm:<6} {cnt:>10}  {cond}")
```

```python
print("Group C")

 
C_names      = [f"C{i}" for i in range(1, 39)]
C_conditions = ["Source ≠ Dealership", "Source = Dealership, DQ < 80%"]
for lf_lbl in ["LF < 10%", "10% ≤ LF < 20%", "20% ≤ LF < 30%",
                "30% ≤ LF < 40%", "40% ≤ LF < 50%", "LF ≥ 50%"]:
    for dp_lbl in ["DP < 0.3%", "0.3% ≤ DP < 0.6%", "0.6% ≤ DP < 0.9%",
                   "0.9% ≤ DP < 1.2%", "1.2% ≤ DP < 1.5%", "DP ≥ 1.5%"]:
        C_conditions.append(
            f"Dealership, DQ ≥ 80%, {lf_lbl}, {dp_lbl}")

 
#  columns
_src = df["Lead Source"].astype(str).str.strip().str.lower().values
_dq  = df["Dealership Quality (DQ)"].astype(float).values
_lf  = df["Lead Flow (LF%)"].astype(float).values
_dp  = df["Dealer Payout% (DP%)"].astype(float).values
 
#  dealership masks
is_dlr = (_src == "dealership")
dlr_ok = is_dlr & (_dq >= 80)
 
# LF
lf_grp = np.where(_lf < 10, 0,
         np.where(_lf < 20, 1,
         np.where(_lf < 30, 2,
         np.where(_lf < 40, 3,
         np.where(_lf < 50, 4, 5))))).astype(np.int32) 

# DP
dp_grp = np.where(_dp < 0.3, 0,
         np.where(_dp < 0.6, 1,
         np.where(_dp < 0.9, 2,
         np.where(_dp < 1.2, 3,
         np.where(_dp < 1.5, 4, 5))))).astype(np.int32)
 
C_idx = np.zeros(n, dtype=np.int32)
C_idx[~is_dlr]              = 0   # C1 — not dealership
C_idx[is_dlr & (_dq < 80)]  = 1   # C2 — dealership DQ < 80%
C_idx[dlr_ok] = 2 + lf_grp[dlr_ok] * 6 + dp_grp[dlr_ok]  # C3–C38

 
# Print result
print(f"  C1  (Not Dealership)     : {(C_idx==0).sum()} customers")
print(f"  C2  (Dealer DQ < 80%)   : {(C_idx==1).sum()} customers")
print(f"  C3-C38 (Dealer DQ >= 80%): {(C_idx>=2).sum()} customers")
print(f"  Unique C states populated: {len(set(C_idx))}/38")
print(f" LF group 2 : {(lf_grp==2).sum()} customers")
print(f" dp_grp: {(dp_grp==2).sum()} customers")


```

```python

#  GROUP D 

print("Group D ")

#Read data from Excel
 
D_names      = ["D1", "D2", "D3", "D4"]
D_conditions = [
    "Branch Ahead of Profitability Target & Behind Disbursal Target",
    "Branch Ahead of Profitability Target & Ahead of Disbursal Target",
    "Branch Behind Profitability Target & Behind Disbursal Target",
    "Branch Behind Profitability Target & Ahead of Disbursal Target",
]
 
# Read the columns
_profit = df["Tracking towards profitability target"].astype(str).values
_disb   = df["Tracking towards disbursal target"].astype(str).values
 
# Assign D 
D_idx = np.full(n, 2, dtype=np.int32)   # 2 = index of D3
D_idx[(_profit == "Ahead")  & (_disb == "Behind")] = 0   # D1
D_idx[(_profit == "Ahead")  & (_disb == "Ahead")]  = 1   # D2
D_idx[(_profit == "Behind") & (_disb == "Ahead")]  = 3   # D4
 
# Print result
print(f"  {'Bucket'} {'Customers'}  Condition")
print(f"  {'─'*60}")
for i, (nm, cond) in enumerate(zip(D_names, D_conditions)):
    print(f"  {nm:<6} {(D_idx==i).sum():>10}  {cond}")
```

```python
print("\nSTEP 6: Group E — Geography")
 
E_names      = [f"E{i}" for i in range(1, 6)]
E_conditions = [
    "Category 1 – Tier 1 Urban",
    "Category 2 – Tier 1 Rural",
    "Category 3 – Tier 2 Urban",
    "Category 4 – Tier 2 Rural",
    "Category 5 – Tier 3 Rural & Others",
]
 
# Read the column
_ecat = df["Branch Competitive Category"].astype(str).str.strip().values
 
# Assign E 
E_idx = np.zeros(n, dtype=np.int32)
for digit, ei in [("1", 0), ("2", 1), ("3", 2), ("4", 3), ("5", 4)]:
    E_idx[pd.Series(_ecat).str.contains(digit, regex=False).values] = ei
 
# Print result
print(f"  {'Bucket':<6} {'Customers':>10}  Condition")
print(f"  {'─'*60}")
for i, (nm, cond) in enumerate(zip(E_names, E_conditions)):
    print(f"  {nm:<6} {(E_idx==i).sum():>10}  {cond}")
```

```python
# STEP 7 — BUILD ONE ROW PER CUSTOMER EXCEL
# =============================================================
print("\nSTEP 7: Building Excel — one row per customer...")
 
# ── Styles ────────────────────────────────────────────────────
HDR_FILL = PatternFill("solid", start_color="1F4E79")
GRP_HDR  = {
    "A": PatternFill("solid", start_color="4472C4"),
    "B": PatternFill("solid", start_color="375623"),
    "C": PatternFill("solid", start_color="843C0C"),
    "D": PatternFill("solid", start_color="7B3F00"),
    "E": PatternFill("solid", start_color="4B0082"),
}
ALT_FILL = {
    "A": PatternFill("solid", start_color="D6E4F7"),
    "B": PatternFill("solid", start_color="E2EFDA"),
    "C": PatternFill("solid", start_color="FCE4D6"),
    "D": PatternFill("solid", start_color="F4CCAC"),
    "E": PatternFill("solid", start_color="E8D5F5"),
}
WHITE  = PatternFill("solid", start_color="FFFFFF")
thin_s = Side(style="thin", color="BFBFBF")
BORDER = Border(left=thin_s, right=thin_s, top=thin_s, bottom=thin_s)
 
def hcell(ws, r, c, v, w=None, fill=None):
    cell = ws.cell(row=r, column=c, value=v)
    cell.font      = Font(bold=True, color="FFFFFF", size=11, name="Arial")
    cell.fill      = fill or HDR_FILL
    cell.alignment = Alignment(horizontal="center", vertical="center", wrap_text=True)
    cell.border    = BORDER
    if w:
        ws.column_dimensions[get_column_letter(c)].width = w
    return cell
 
def dcell(ws, r, c, v, fill=None, bold=False, align="center", fmt=None):
    cell = ws.cell(row=r, column=c, value=v)
    cell.font      = Font(bold=bold, size=10, name="Arial")
    cell.fill      = fill or WHITE
    cell.alignment = Alignment(horizontal=align, vertical="center", wrap_text=True)
    cell.border    = BORDER
    if fmt:
        cell.number_format = fmt
    return cell
 
wb = Workbook()
ws = wb.active
ws.title = "A-E Customer Conditions"
ws.freeze_panes = "B2"
ws.row_dimensions[1].height = 35
 
# ── Header row ────────────────────────────────────────────────
# Column layout:
# 1=Customer | 2=A Cat | 3=A Condition | 4=B Cat | 5=B Condition
#            | 6=C Cat | 7=C Condition | 8=D Cat | 9=D Condition
#            | 10=E Cat | 11=E Condition
 
headers = [
    (1,  "Customer",    22,  None),
    (2,  "A Category",  12,  "A"),
    (3,  "A Condition", 28,  "A"),
    (4,  "B Category",  12,  "B"),
    (5,  "B Condition", 58,  "B"),
    (6,  "C Category",  12,  "C"),
    (7,  "C Condition", 58,  "C"),
    (8,  "D Category",  12,  "D"),
    (9,  "D Condition", 50,  "D"),
    (10, "E Category",  12,  "E"),
    (11, "E Condition", 28,  "E"),
]
 
for ci, label, width, grp in headers:
    fill = GRP_HDR[grp] if grp else HDR_FILL
    hcell(ws, 1, ci, label, w=width, fill=fill)
 
# ── Data rows — one row per customer ─────────────────────────
for row_i in range(n):
    ri   = row_i + 2
    # alternate row color — use A group color as base
    fill = ALT_FILL["A"] if row_i % 2 == 0 else WHITE
 
    ai = A_idx[row_i]
    bi = B_idx[row_i]
    ci = C_idx[row_i]
    di = D_idx[row_i]
    ei = E_idx[row_i]
 
    values = [
        (1,  df["Name of Customer"].iloc[row_i],  "left",   fill),
        (2,  A_names[ai],                          "center", ALT_FILL["A"] if row_i%2==0 else WHITE),
        (3,  A_conditions[ai],                     "left",   ALT_FILL["A"] if row_i%2==0 else WHITE),
        (4,  B_names[bi],                          "center", ALT_FILL["B"] if row_i%2==0 else WHITE),
        (5,  B_conditions[bi],                     "left",   ALT_FILL["B"] if row_i%2==0 else WHITE),
        (6,  C_names[ci],                          "center", ALT_FILL["C"] if row_i%2==0 else WHITE),
        (7,  C_conditions[ci],                     "left",   ALT_FILL["C"] if row_i%2==0 else WHITE),
        (8,  D_names[di],                          "center", ALT_FILL["D"] if row_i%2==0 else WHITE),
        (9,  D_conditions[di],                     "left",   ALT_FILL["D"] if row_i%2==0 else WHITE),
        (10, E_names[ei],                          "center", ALT_FILL["E"] if row_i%2==0 else WHITE),
        (11, E_conditions[ei],                     "left",   ALT_FILL["E"] if row_i%2==0 else WHITE),
    ]
 
    for col, val, align, f in values:
        dcell(ws, ri, col, val, fill=f, align=align)
 
# ── Save ──────────────────────────────────────────────────────
wb.save(OUT_EXCEL)
print(f"\n  Excel saved → {OUT_EXCEL}")
print(f"  Rows: {n} customers | Columns: 11")
```
